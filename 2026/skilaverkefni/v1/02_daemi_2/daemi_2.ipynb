Eftirfarandi biti $OABC$ ber álag í $xy$ og $zx$ plönum. Gerið ráð fyrir að gefnar stærðir á mynd hafi a.m.k þrjá markverða tölustafi eða fleiri sé svo sýnt.

(a) Setjið upp Macauley föll fyrir álag, skerkraft og beygjuvægi í $xy$ og $xz$ plönum.
(b) Finnið undirstöðukrafta í $y$ og $z$ stefnur í punktum $O$ og $C$.
(c) Teiknið sker- og vægisferla fyrir bæði $xy$ og $xz$ plön. Berið kennsl á krístíska punkta fyrir skerkraft og beygjuvægi.
(d) Finnið hæstu beygjuspennur (tog og þrýsti) m.v. gefið þversnið og merkið viðeigandi staðsetningar í þvernsiði.

## Lausn

### (a) Macauley föll
Skoðum $xy$ og $xz$ plönin hvort fyrir sig og setjum upp Macauley föll fyrir viðeigandi stefnur.

Við munum bregða útaf þeim formerkjavenjum sem notaðar eru í Shigley. Formerki á kröftum og vægjum eru miðuð við stefnu þeirra í skilgreindu hnitakerfi. Til þess að það gangi upp, þá eru skerkraftar og vægi skigreind á þversniði bita sem hefur normal í $+x$ stefnu, þ.e. við skerum í bita í punkti $x$ og skoðum krafta og vægi á því þversniði sem er vinstra megin (með normal í $+x$). Formerki á hinu þversniðinu með normal í $-x$ hefur andstæð formerki vegna þriðja lögmáls Newtons.

Látum kraftinn í punkti A vera $\vec{P}_A = [0, 0, -P_{Az}]^T$ með $P_{Az}=1.5 \text{ kN}$ og jafndreifða álagið milli $BC$ vera $\vec{w}_{BC} = [0, -w\cos\theta, -w\sin\theta]^T$ með $w=2.00 \text{ kN}$ og $\theta = 30^\circ$.

#### Macauley föll í $xy$ plani
Álagið er $q_y(x)$
$$
    q_y(x) =
        R_{1y} \langle x \rangle^{-1}
        - w \cos\theta \langle x - x_B \rangle^0
        + w \cos\theta \langle x - x_C \rangle^0
        + R_{2y} \langle x - x_C \rangle^{-1}.
$$

Innri skerkraftur $V_y(x)$ á þversnið í punkti $x$ með normal í $+x$. Við heildum $q_y(x)$ einu sinni með tilliti til $x$ og víxlum formerkjum því jákvæður ytri kraftur veldur neikvæðum innri skerkrafti.
$$
    V_y(x) =
        - R_{1y} \langle x \rangle^0
        + w \cos\theta \langle x - x_B \rangle^1
        - w \cos\theta \langle x - x_C \rangle^1
        - R_{2y} \langle x - x_C \rangle^0.
$$

Að lokum er innra vægið $M_z(x)$ í þversniði í punkti $x$ með normal í $+x$ stefnu fengið með því að heilda $V_y(x)$ einu sinni með tilliti til $x$ og setja formerki miðað við hægri handarreglu og punkt $x$ á bitanum. Athugum að ytri kraftur í $+y$ stefnu myndar neikvætt ytra vægi um $z$ ás í punkti $x$ sem veldur þ.a.l jákvæðu innra vægi um $z$ ás í punkti $x$, og öfugt, skv. hægri-handar reglu.
$$
    M_z(x) 
        =
        R_{1y} \langle x \rangle^1
        - \frac{w \cos\theta}{2} \langle x - x_B \rangle^2
        + \frac{w \cos\theta}{2} \langle x - x_C \rangle^2
        + R_{2y} \langle x - x_C \rangle^1
$$

#### Macauley föll í $xz$ plani
Gerum það sama fyrir $xz$ planið. Athugum hér fyrir vægið $M_y(x)$ að ytri kraftur í $+z$ stefnu myndar jákvætt vægi um $y$ ás í punkti $x$ sem veldur þ.a.l neikvæðu innra vægi um $y$ ás í punkti $x$, og öfugt, skv. hægri-handar reglu.
$$
    q_z(x) 
    = 
        R_{1z} \langle x \rangle^{-1}
        - P_{Az} \langle x - x_A \rangle^{-1}
        - w \sin\theta \langle x - x_B \rangle^0
        + w \sin\theta \langle x - x_C \rangle^0
        + R_{2z} \langle x - x_C \rangle^{-1}.
$$
$$
    V_z(x) 
    =
        - R_{1z} \langle x \rangle^0
        + P_{Az} \langle x - x_A \rangle^0
        + w \sin\theta \langle x - x_B \rangle^1
        - w \sin\theta \langle x - x_C \rangle^1
        - R_{2z} \langle x - x_C \rangle^0.
$$
$$
    M_y(x) 
    =
        - R_{1z} \langle x \rangle^1
        + P_{Az} \langle x - x_A \rangle^1
        + \frac{w \sin\theta}{2} \langle x - x_B \rangle^2
        - \frac{w \sin\theta}{2} \langle x - x_C \rangle^2
        - R_{2z} \langle x - x_C \rangle^1.
$$

### Undirstöðukraftar í punktum $O$ og $C$
Fyrir $xy$ planið þá skoðum við $V_y(x_C)$ til að setja upp kraftajafnvægi í $y$ stefnu $V_y(x_C) = 0$
$$
\begin{aligned}
    V_y(x_C) 
    &=
        - R_{1y}
        + w \cos\theta (x_C - x_B)
        - w \cos\theta (x_C - x_C)
        - R_{2y} \\
    &= 
        - R_{1y}
        + w \cos\theta (x_C - x_B)
        - R_{2y}
    = 0.
\end{aligned}
$$

Fyrir vægisjafnvægi skoðum við $M_z(x_C) = 0$ sem er jafngilt því að skoða vægisjafnvægi um punkt $C$.
$$
\begin{aligned}
    M_z(x_c) 
    &= 
        R_{1y}x_C
        - \frac{w \cos\theta}{2} (x_C - x_B)^2
        + \frac{w \cos\theta}{2} (x_C - x_C)^2
        + R_{2y} ( x_C - x_C ) \\
    &= 
        R_{1y}x_C
        - \frac{w \cos\theta}{2} (x_C - x_B)^2
    = 0.
\end{aligned}
$$

Vægisjafnvægi gefur að
$$
    R_{1y} = \frac{w \cos\theta}{2x_C} ( x_C - x_B )^2.
$$

Þá fáum við úr kraftajafnvægi að 
$$
\begin{aligned}
    R_{2y} 
    &= 
        - R_{1y}
        + w \cos\theta (x_C - x_B) \\
    &=
        -\frac{w \cos\theta}{2x_C} ( x_C - x_B )^2
        + w \cos\theta (x_C - x_B).
\end{aligned}
$$

Úr $xz$ plani fáum við svipað með því að setja $V_z(x_C)=0$ og $M_y(x_C)=0$
$$
\begin{aligned}
    V_z(x_C) 
    &=
        - R_{1z}
        + P_{Az}
        + w \sin\theta (x_C - x_B)
        - w \sin\theta (x_C - x_C)
        - R_{2z}\\
    &=
        - R_{1z}
        + P_{Az}
        + w \sin\theta (x_C - x_B)
        - R_{2z}
    = 0.
\end{aligned}
$$

$$
\begin{aligned}
    M_y(x) 
    &=
        - R_{1z} x_C
        + P_{Az} (x_C - x_A)
        + \frac{w \sin\theta}{2} (x_C - x_B)^2
        - \frac{w \sin\theta}{2} (x_C - x_C)^2
        - R_{2z} (x_C - x_C) \\
    &=
        - R_{1z} x_C
        + P_{Az} (x_C - x_A)
        + \frac{w \sin\theta}{2} (x_C - x_B)^2
    = 0.
\end{aligned}
$$

Vægisjafnvægi gefur 
$$
    R_{1z}
    =  
        \frac{1}{x_C}
        \left( 
            P_{Az} (x_C - x_A)
        + \frac{w \sin\theta}{2} (x_C - x_B)^2
        \right).
$$

Kraftajafnvægi gefur
$$
\begin{aligned}
    R_{2z} 
    &=
        -R_{1z}
        + P_{Az}
        + w \sin\theta (x_C - x_B)
\end{aligned}
$$

```python
#| echo: true
#| output: true

# reiknum undirstöðukrafta
import numpy as np

# gefnar stærðir
theta = 30*np.pi/180 # rad
P_Az = 1500 # N
w = 2000 # N

# hnit punkta
x_0 = 0
x_A = 0.5
x_B = x_A + 1
x_C = x_B + 1.5

# undirstöðukraftar í y stefnu
R_1y = (
    (w * np.cos(theta)) 
    * (x_C - x_B)**2 
    / (2 * x_C)
)

R_2y = (
    -R_1y
    + w * np.cos(theta) * (x_C - x_B)
)
# undirstöðukraftar í z stefnu
R_1z = (
    (
        P_Az * (x_C - x_A) 
        + w * np.sin(theta) 
        * (x_C - x_B)**2 / 2
    )
    / x_C
)

R_2z = (
    - R_1z
    + P_Az
    + w * np.sin(theta) * (x_C - x_B)
)

# söfnum saman
R_1 = np.array([0, R_1y, R_1z])
R_2 = np.array([0, R_2y, R_2z])

# prentum
print(f"Undirstöðukr. í punkti 1, R1=(R_1x, R_1y, R_1z)=(0, {R_1y:.2f}, {R_1z:.2f}) N")
print(f"Undirstöðukr. í punkti 2, R2=(R_2x, R_2y, R_2z)=(0, {R_2y:.2f}, {R_2z:.2f}) N")

# athugum niðurstöðu m.v. jöfnur fyrir krafta og vægi
# skilgreinum fyrst einvítt macauley fall
mac = lambda x, a, n: 0 if (n < 0) or (x < a) else (x - a) ** n

# skilgreinum V(x) og M(x) sem vigurföll
def V_fun(x, R_1, R_2): 
    Vy = (
        - R_1[1] * mac(x, x_0, 0)
        + w * np.cos(theta) * mac(x, x_B, 1)
        - w * np.cos(theta) * mac(x, x_C, 1)
        - R_2[1] * mac(x, x_C, 0)
    )

    Vz = (
        - R_1[2] * mac(x, x_0, 0)
        + P_Az * mac(x, x_A, 0)
        + w * np.sin(theta) * mac(x, x_B, 1)
        - w * np.sin(theta) * mac(x, x_C, 1)
        - R_2[2] * mac(x, x_C, 0)
    )

    V = np.array(
        [0, Vy, Vz]
    )
    return V

def M_fun(x, R_1, R_2): 
    My = (
        - R_1[2] * mac(x, x_0, 1)
        + P_Az * mac(x, x_A, 1)
        + w * np.sin(theta) * mac(x, x_B, 2) / 2
        - w * np.sin(theta) * mac(x, x_C, 2) / 2
        - R_2[2] * mac(x, x_C, 1)
    )

    Mz = (
        + R_1[1] * mac(x, x_0, 1)
        - w * np.cos(theta) * mac(x, x_B, 2) / 2
        + w * np.cos(theta) * mac(x, x_C, 2) / 2
        + R_2[1] * mac(x, x_C, 1)
    )

    M = np.array(
        [0, My, Mz]
    )
    return M

# reiknum V(xC) og M(xC) tékkum að allt sé =0
print(f"Kraftajafnvægi V(x_C)= {V_fun(x_C, R_1, R_2)} N")
print(f"Vægisjafnvægi M(x_C)= {M_fun(x_C, R_1, R_2)} Nm")
```

### (a) Macauley föll - önnur aðferð
Við getum nýtt okkur kosti þess að formerki krafta og vægja fylgja því hnitakerfi sem er skilgreint og skrifað upp Macauley föllin með þrívíðum vigrum í $xyz$ plani. Með því getum við soðið niður þessi sex Macauley föll úr fyrri aðferð niður um helming í þrjú föll.

Verkefnið er aðeins einfaldað af þeim sökum að bitinn liggur allur á $x$ ásnum og allir kraftar liggja í $yz$ planinu, þ.e. engir áskraftar í $\pm x$ stefnur.

Látum undirstöðurkrafta vera vigra $\vec{R}_1$ og $\vec{R}_2$. Látum einnig álagið í punkti $A$ vera $\vec{P}_A = [0, 0, P_{Az}]^T$ með $P_{Az}=-1.5 \text{ kN}$ og jafndreifða álagið milli $BC$ vera $\vec{w}_{BC} = [0, w\cos\theta, w\sin\theta]^T$ með $w=-2.00 \text{ kN/m}$ og $\theta = 30^{\circ}$. Athugið að hér látum við stefnu vigrana vera stjórnað af formerki skalargildanna sem mynda vigranna og setjum ekki formerki inn í vigrana sjálfa.

Álagið $\vec{q}$ má þá rita sem
$$
    \vec{q}(x) 
    =
    \vec{R}_1 \langle x \rangle^{-1}
    + \vec{P}_A \langle x - x_A \rangle^{-1}
    + \vec{w}_{BC} \langle x - x_B \rangle^0
    - \vec{w}_{BC} \langle x - x_C \rangle^0
    + \vec{R}_2 \langle x - x_C \rangle^{-1}.
$$

Innri skerkrafturinn $\vec{V}(x)$ fæst með því að heilda Macauley föllin í $\vec{q}(x)$ einu sinni og víxla formerkjum vigranna
$$
    \vec{V}(x)
    =
    - \vec{R}_1 \langle x \rangle^0
    - \vec{P}_A \langle x - x_A \rangle^0
    - \vec{w}_{BC} \langle x - x_B \rangle^1
    + \vec{w}_{BC} \langle x - x_C \rangle^1
    - \vec{R}_2 \langle x - x_C \rangle^0.
$$

Innra vægið $\vec{M}$ þarf að nálgað aðeins öðruvísi en áður. Er það sökum þess að vægið vegna ytri krafts $\vec{F}$ er krossmargfeldi stefnuvigurs, frá punkti $x$ í átakspunkt kraftsins, og kraftavigurisins $\vec{F}$. Við getum því ekki heildað eingöngu Macauley föllin í $\vec{V}(x)$ einu sinni til að fá út $\vec{M}(x)$, því $\vec{V}(x)$ lýsir innri kröftum en ekki ytri kröftum. Við getum þó víxlað aftur formerkjum vigranna í $\vec{V}(x)$ til að fá ytri kraftana og svo notað þá til að finna vægið.

Fyrir ákveðinn kraft $F$ sem verkar í punkti $p$ þá er stefnuvigur í þann punkt úr punkti $x$ með $x \geq p$ vigurinn $\vec{r} = (p - x)\hat{i} = -(x - p)\hat{i} = -\langle x - p \rangle^1 \hat{i}$ þar sem $\hat{i}$ er einingavigurinn samsíða $x$ ás.

Þar með getum við smíðað fallið $\vec{M}(x)$ fyrir innra vægið í punkti $x$ sem
$$
\begin{aligned}
    \vec{M}(x)
    &=
    \left ( -\langle x \rangle^1 \hat{i} \times \vec{R}_1 \langle x \rangle^0 \right) \\
    &+ \left( -\langle x - x_A \rangle^1 \hat{i} \times \vec{P}_A \langle x - x_A \rangle^0 \right) \\
    &+ \left( -\frac{1}{2} \langle x - x_B \rangle^1 \hat{i} \times \vec{w}_{BC} \langle x - x_B \rangle^1 \right) \\
    &- \left( -\frac{1}{2} \langle x - x_C \rangle^1 \hat{i} \times \vec{w}_{BC} \langle x - x_C \rangle^1 \right) \\
    &+ \left( -\langle x - x_C \rangle^1 \hat{i} \times \vec{R}_2 \langle x - x_C \rangle^0 \right). \\
\end{aligned}
$$

Það er engin þörf á því að nota Macauley föll á bæði stefnuvigrana og kraftana. Það dugar að nota Macauley föll á kraftana, því allt krossmargfeldið verður núllvigurinn svo lengi sem annar hvor vigranna er núllvigurinn. Hins vegar hef ég Macauley föllin á báðum vigrum til að hafa samanburð við einvíðu aðferðina hér fyrir ofan. Í stað þess að heilda og hækka veldið á Macauley föllunum um einn, þá höfum við frekar margfaldað hvert Macauley fall við sjálft sig í gegnum krossmargfeldið, sem í raun hækkar samanlagðan veldisvísi um einn. Formerkin fá einnig að halda sér m.v. stefnur vigranna og munu krossmargfeldin sjá um að útkoman hafi rétt formerki miðað við hnitakerfið.

Hægt væri að losna við krossmargfeldin með því að skoða vigrana og beita hægri-handar reglu - sem var einmitt það sem var notað í fyrri útgáfu lausnar. Þessi krossmargfeldi má frekar forrita og reikna tölulega. Kosturinn hér er sá að formerki á vægi um $y$ og $z$ ás koma út úr krossmargfeldunum frekar en frá okkur sjálfum við beitingu hægri-handar reglu.

Við leysum fyrir undirstöðurkrafta sem fyrr með því að setja $\vec{V}(x_C) = 0$ og $\vec{M}(x_C)=0$, sem gefur
$$
    - \vec{R}_1
    - \vec{P}_A
    - \vec{w}_{BC} (x_C - x_B)
    - \vec{R}_2
    =
    0
$$
og
$$
\begin{aligned}
    x_C \hat{i} \times \vec{R}_1 
    =
    & \left( -(x_C - x_A) \hat{i} \times \vec{P}_{A} \right) \\
    &+ \left( -\frac{1}{2} (x_C - x_B)\hat{i} \times \vec{w}_{BC} (x_C - x_B) \right)
\end{aligned}
$$
Þar sem $\vec{R}_1 = [0, R_{1y}, R_{1z}]^T$ og $\hat{i}=[1, 0, 0]^T$ þá er $\hat{i} \times \vec{R}_1 = [0, -R_{1z}, R_{1y}]^T$. Við getum skalað fyrri vigurinn með $x_C$ og fengið $x_C\hat{i} \times \vec{R}_1= x_C[0, -R_{1z}, R_{1y}]^T$. Þar með fæst lausnin
$$
\begin{aligned}
    \begin{bmatrix}
        0 \\
        -R_{1z} \\
        R_{1y}
    \end{bmatrix}
    &=
    \frac{1}{x_C}
    \left[
        \left( -(x_C - x_A) \hat{i} \times \vec{P}_{A} \right)
        + \left( -\frac{1}{2} (x_C - x_B)\hat{i} \times \vec{w}_{BC} (x_C - x_B) \right)
    \right] \\
    &= 
    \frac{1}{x_C}
    \left[
        \left( -(x_C - x_A) \hat{i} \times \vec{P}_{A} \right)
        + \left( -\frac{1}{2} (x_C - x_B)\hat{i} \times \vec{w}_{BC} (x_C - x_B) \right)
    \right]
\end{aligned}  
$$

Við getum lagað til hnitin í lausninni til að stemma við $\vec{R}_1 = [0, R_{1y}, R_{1z}]$ og leyst svo fyrir $\vec{R}_2$
$$
    \vec{R}_2
    = 
    - \vec{R}_1
    - \vec{P}_A
    - \vec{w}_{BC} (x_C - x_B)
$$

```python
#| echo: true
#| output: true

import numpy as np

# skilgreinum einvítt Macauley fall
mac = lambda x, a, n: 0 if (n < 0) or (x < a) else (x - a) ** n

# skilgreinum gefnar stærðir
P_A = np.array([0, 0, -1500]) # N
w = 2000 # N/m
theta = 30 * np.pi / 180 # radians
w_BC = np.array([0, -w*np.cos(theta), -w*np.sin(theta)]) # N/m

# einingavigur í x
vec_i = np.array([1, 0, 0])

# skilgreinum V(x) og M(x) sem föll
def V_fun(x, R_1, R_2): 
    V = (
        - R_1 * mac(x, x_0, 0)
        - P_A * mac(x, x_A, 0)
        - w_BC * mac(x, x_B, 1)
        + w_BC * mac(x, x_C, 1)
        - R_2 * mac(x, x_C, 0)
    )
    return V

def M_fun(x, R_1, R_2):
    M = (
        + np.cross(mac(x, x_0, 1) * vec_i, R_1 * mac(x, x_0, 0))
        + np.cross(mac(x, x_A, 1) * vec_i, P_A * mac(x, x_A, 0))
        + np.cross(0.5 * mac(x, x_B, 1) * vec_i, w_BC * mac(x, x_B, 1))
        - np.cross(0.5 * mac(x, x_C, 1) * vec_i, w_BC * mac(x, x_C, 1))
        + np.cross(+mac(x, x_C, 1) * vec_i, R_2 * mac(x, x_C, 0,))

    )
    return M

# reiknum undirstöðukrafta skv V(x) og M(x)
R_1 = (
    (
        + np.cross(-(x_C - x_A) * vec_i, P_A)
        + np.cross(-0.5 * (x_C - x_B) * vec_i, w_BC * (x_C - x_B))
    )
    / x_C
)

# víxlum til í R_1
R_1 = np.array([0, R_1[2], -R_1[1]])

# leysum fyrir R_1
R_2 = -R_1 - P_A - w_BC * (x_C - x_B)

print(f"Prentum niðurstöður og staðfestum jafnvægi V(x_C)=0 og M(x_C)=0")
print(f"R_1 = {R_1} N")
print(f"R_2 = {R_2} N")
print(f"V_fun(x_C, R_1, R_2) = {V_fun(x_C, R_1, R_2)} N")
print(f"M_fun(x_C, R_1, R_2) = {M_fun(x_C, R_1, R_2)} Nm")
```

### (c) Sker- og vægisferlar
Teiknum ferla með python m.v. skilgreind föll fyrir $\vec{V}(x)$ og $\vec{M}(x)$ úr lið (b).

### (d) Hæstu beygjuspennur
Einungis eru skerkraftar og beygjuvægi í þessu verkefni, engin vinduvægi. Þar af leiðandi verður bitinn einugis fyrir normalspennu vegna beygju (beygjuspennu) og skerspennu vegna skerkrafta þar sem ekki er um hreina beygju (e. pure bending) að ræða.

Heildarbeygjuvægi í $yz$ þversniði fæst með því að finna þá staðsetningu þar sem normið af vægisvigrinum tekur hæsta gildi. Vigurnormið er 
$$
    \|\vec{M}(x)\|
    =
    \sqrt{ M_y(x)^2 + M_z(x)^2 }.
$$

Þar sem normið er skalargilt fall af $x$ þá getum við fundið hámarkið með því að diffra og finna núllstöðvar. Best er að gera það tölulega með $\texttt{scipy.optimize}$ pakkanum í python og fallið $\texttt{minimize\_scalar}$ sem leitar að lágmarki falls. Við getum þá leyst fyrir hámarkið með því að finna lágmark á neikvæða norminu, þ.e. $-\|\vec{M}(x)\|$.

```python
#| echo: true
#| output: true

# finnum hámarksgildi á normi M og samsvarandi punkt x
from scipy.optimize import minimize_scalar

# fallið minimize_scalar finnur lágmark svo við finnum lágmark -M_norm
res = minimize_scalar(
    lambda x: -np.linalg.norm(M_fun(x, R_1, R_2)),
    bounds=(0, x_C),
    method='bounded'   
)
x_max = res.x
M_max = M_fun(x_max, R_1, R_2)
print(f"Hámörk beygjuvægis eru í x={x_max:.3f} m")
print(f"M(x_max)={M_max} Nm og ||M||={np.linalg.norm(M_max):.2f} Nm")
print(f"Hæsta vægi um y ás er M_y={M_max[1]:.3f} Nm")
print(f"Hæsta vægi um z ás er M_z={M_max[2]:.2f} Nm")

# teiknum saman M(x), norm M(x) og x_max á einni mynd
plt.figure(figsize=(10, 6))
plt.plot(x_vals, M_vals[:, 2], label='M_z(x)')
plt.plot(x_vals, M_vals[:, 1], label='M_y(x)')
plt.plot(
    x_vals, 
    np.linalg.norm(M_vals, axis=1), 
    label='||M(x)||',
    linestyle='--', 
    color='black'
)
plt.axvline(x_max, color='red', linestyle=':', label='x_max')
plt.title('Beygjuvægi M(x) og norm M(x)')
plt.xlabel('x (m)')
plt.ylabel('Beygjuvægi (Nm)')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

Til að finna spennur í þversniðinu og afstöðu núllássins þurfum við að reikna flatartregðurnar um $y$ og $z$ ásana. Munum að flatartregða fyrir rétthyrnt þversnið er $I=bh^3/12$. Finnum einnig halla núllássins í krítískum punkti $x_{max}$ með
$$
    \tan\phi
    =
    \frac{M_yI_z}{M_zI_y},
$$
þar sem $\phi$ er reiknað jákvætt fyrir snúning frá $+z$ ás að $+y$ ás.

```python
#| echo: true
#| output: true

# finnum flatarmiðju og flatartregður um y og z ás
# lóðréttur flangsi (#1)
b_1 = 22.225e-3 # m
h_1 = 63.5e-3 # m
y_1 = h_1/2 # m
A_1 = b_1 * h_1 # m^2
I1_y = (h_1 * b_1**3) / 12 # m^
I1_z = (b_1 * h_1**3) / 12 # m^4

# láréttur flangsi (#2)
b_2 = 101.600e-3 # m
h_2 = 22.225e-3 # m
y_2 = h_2/2 + h_1 # m
A_2 = b_2 * h_2 # m^2
I2_y = (h_2 * b_2**3) / 12 # m^4
I2_z = (b_2 * h_2**3) / 12 # m^4

# flatarmiðja (e. centroid)
y_c = (A_1*y_1 + A_2*y_2) / (A_1 + A_2)
print(f"Flatarmiðja y_c = {y_c*1e3:.2f} mm frá neðri brún")

# flatartregður Iy og Iz
I_y = I1_y + I2_y
I_z = (
    I1_z + A_1 * (y_c - y_1) ** 2 
    + I2_z + A_2 * (y_c - y_2) ** 2
)
print(f"Flatartregða Iy = {I_y*1e12:.3e} m^4")
print(f"Flatartregða Iz = {I_z*1e12:.3e} m^4")

# halli núllássins í krítískum punkt x_max er
phi = np.atan(M_fun(x_max, R_1, R_2)[1]*I_z/(M_fun(x_max, R_1, R_2)[2]*I_y))
print(f"Halli núllássins phi = {np.degrees(phi):.2f}° frá +z ás") 
```

Til þess að reikna spennur í krítískum punktum getum við notað jöfnuna
$$
    \sigma(y,z) = \frac{M_{max,y}z}{I_z} - \frac{M_{max,z}y}{I_y},
$$
þar sem við þurfum að setja inn hnitin á krítískum punktum $(y,z)$ í þversniðinu. Þessi framsetning er best ef við þurfum að reikna í höndunum með reiknivél.

Við getum líka nálgast þetta með vigrum og reiknað með python.

Hæstu tog- og þrýstispennur fáum við í þeim punktum í þversniðinu sem eru lengst frá núllásnum, þ.e. hafa lengstu ofanvörp á normalvigur núllássins, annað hvort í jákvæða eða neikvæða stefnu. Jákvæður normall $\vec{n}$ á núllásinn er einingarvigurinn $\hat{j}$ eftir $y$ ás sem hefur verið snúið um hornið $\phi$.
$$
    \vec{n}
    =
    \begin{bmatrix}
        0 \\
        \cos\phi \\
        -\sin\phi
    \end{bmatrix}.
$$

Fyrir punkt $\vec{p}=[y, z]^T$ í þversniðinu er ofanvarpið á normalinn $\vec{n}$ gefið með innfeldinu
$$
    p_n
    =
    \vec{p} \cdot \vec{n}
    =
    y \cos\phi
    - z \sin\phi.
$$

Teiknum þversnið með núllás og staðsetningu hámarks spennu til að sýna staðsetningu hámarks tog- og þrýstispennu.

![](fig/v1_d2_d1.png){width=60% fig-align="center"}

Spennurnar finnum við svo með því að taka krítískan vægisvigur $\vec{M}_{max}$ og deila í hvert hnit með viðeigandi flatartregðu og reikna svo krossfeldi þess vigurs og $p_n\vec{n}$. Jafngild fylkjaaðferð er að margalda vigurinn $\vec{M}_{max}$ frá vinstri með fylkinu $diag(\vec{I})^{-1}$ sem er andhverfa fylkis sem hefur flatartregðurnar $I_x$, $I_y$ og $I_z$ á hornalínunni og reikna krossmargfeldi við $p_n \vec{n}$. Setjum $I_x=\infty$ því $yz$ þversnið bitans hefur ekki skilgreint flatartregðuvægi um $x$ ás, því $x$ ásinn er normall á $yz$ þversniðið. Þannig fæst
$$
    \sigma_p
    =
    \left(
    diag(\vec{I})^{-1}
    \vec{M}_{max}
    \right)
    \times
    p_n \vec{n}.
$$
Með þessu fáum við stærð spennunar og formerki hennar í einu skrefi - sem segir til um hvort um togspennu (+) eða þrýstispennu (-) er að ræða.

```python
#| echo: true
#| output: true

# reiknum hæstu spennur í þversniðinu

# skilgreinum krítíska punkta
p_C = np.array([0, h_1 + h_2/2 - y_c, b_2/2])
p_A = np.array([0, -y_c, -b_1/2])

# reiknum jákvæðan normal
norm_vec = np.array([0, np.cos(phi), -np.sin(phi)])

# fylki með flatartregður á hornalínunni
diag_I = np.diag([np.inf, I_y, I_z]) # setjum Ix=inf (er óþarfi)

# reiknum hæstu spennur í krítískum punktum A og C
MI_vec = np.linalg.inv(diag_I) @ M_fun(x_max, R_1, R_2) 

# reiknum spennur í punktum A og C 
sigma_C = np.cross(MI_vec, np.dot(p_C, norm_vec) * norm_vec)
sigma_A = np.cross(MI_vec, np.dot(p_A, norm_vec) * norm_vec)

print(f"Spenna í punkti A er = {sigma_A/1e6} MPa")
print(f"Spenna í punkti C er = {sigma_C/1e6} MPa")
```

Í punkti A er því togspenna $\sigma_A = 34.5 \text{ MPa}$ og í punkti C er þrýstispennan $\sigma_C = -32.0 \text{ MPa}$. Engin skerspenna er í báðum þessum punktum því þeir eru á ytra yfirborði þversniðsins.

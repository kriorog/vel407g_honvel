## Lausn

### (a) Undirstöðukraftar og vægi í punkti A
Nálgumst þetta verkefni með vigrum m.v. skilgreint hnitakerfi á mynd.

Höfum tvo krafta í punktum $C$ og $E$, sem eru $\vec{F}_C = [F_{Cx}, 0, 0]^T$  og $\vec{F}_E = [0, F_{Ey}, 0]^T$ þar sem $F_{Cx} = 300 \text{ N}$ og $F_{Ey} = 400 \text{ N}$. Kraftajafnvægi gefur þá fyrir undirstöðukraft $\vec{R}_A$ í punkti A
$$
    \vec{R}_A
    + \vec{F}_C
    + \vec{F}_E
    = \vec{0},
$$
og þar með
$$
    \vec{R}_A
    =
    - \vec{F}_C
    - \vec{F}_E
    =
    \begin{bmatrix}
        -300 \\
        -400 \\
        0
    \end{bmatrix} \text{ N}.
$$

Fyrir vægisjafnvægi þurfum við stefnuvigra frá punkti $A$ í átakspunkta ytri kraftanna. Þar sem $A$ er upphafspunktur hnitakerfisins $\vec{p}_A = [0, 0, 0]^T$ þá eru stefnuvigrarnir jafnir punktunum $\vec{p}_C$ og $\vec{p}_E$. Vægisjafnvægi um punkt $A$ er þá
$$
\begin{aligned}
    \vec{M}_A
    + (\vec{p}_A - \vec{p}_A) \times \vec{R}_A
    + (\vec{p}_C - \vec{p}_A) \times \vec{F}_C
    + (\vec{p}_E - \vec{p}_A) \times \vec{F}_E
    = 0,
\end{aligned}
$$
og þar með
$$
    \vec{M}_A
    =
    - (\vec{p}_C \times \vec{F}_C)
    - (\vec{p}_E \times \vec{F}_E)
$$

```python
import numpy as np

# reiknum undirstöðukrafta og vægi í punkti A

# skilgreinum fyrst punktana A, C, E í metrum
p_A = np.array([0, 0, 0]) # m
p_C = np.array([55 + 40, 0, 50]) / 1e3 # m
p_E = np.array([55, 0, -60]) / 1e3 # m
p_D = np.array([55 + 40, 0, 0]) / 1e3 # m

# skilgreinum kraftana í hverjum punkti í N
F_C = np.array([300, 0, 0]) # N
F_E = np.array([0, 200, 0]) # N

# reiknum undirstöðukrafta í punkti A
R_A = - F_C - F_E
print(f"Undirstöðukraftur í punkti A er R_A = {R_A} N")

# reiknum vægi um punkt A
M_A = -np.cross(p_C, F_C) - np.cross(p_E, F_E)
print(f"Undirstöðuvægi í punkti A er M_A = {M_A} Nm")

# athugum jafnvægi krafta og vægis
sum_F = R_A + F_C + F_E
sum_M = (
    M_A 
    + np.cross(p_C, F_C) 
    + np.cross(p_E, F_E)
)
print(f"Samtala krafta er sum(F) = {sum_F} N (á að vera 0)")
print(f"Samtala vægis um punkt A er sum(M) = {sum_M} Nm (á að vera 0)")    
```

Þar með er undirstöðukraftur og vægi í punkti A
$$
    \vec{R}_A
    =
    \begin{bmatrix}
        -300 \\
        -400 \\
        0
    \end{bmatrix} \text{ N},
$$
og
$$
    \vec{M}_A
    =
    \begin{bmatrix}
        -12 \\
        -15 \\
        -11
    \end{bmatrix} \text{ Nm}.
$$  

### (b) Sker- og vægisferlar
Til að teikna sker- og vægisföll þá getum við sett upp Macauley föll fyrir álag, innri skerkraft og innra vægi.

Ytra álag $\vec{q}(x)$ er
$$
\begin{aligned}
    \vec{q}(x)
    &=
    \vec{M}_A \langle x \rangle^{-2}
    + \vec{R}_A \langle x \rangle^{-1}
    + \vec{F}_C \langle x - x_C \rangle^{-1}
    + \vec{F}_E \langle x - x_E \rangle^{-1}.
\end{aligned}
$$

Innriskerkraftur $\vec{V}(x)$ er (með réttum formerkjum mv. hnitakerfi) þar sem jákvæðir ytri kraftar og vægi valda neikvæðum innri kröftum og vægjum
$$
\begin{aligned}
    \vec{V}(x)
    &=
    - \vec{M}_A \langle x \rangle^{-1}
    - \vec{R}_A \langle x \rangle^{0}
    - \vec{F}_C \langle x - x_C \rangle^{0}
    - \vec{F}_E \langle x - x_E \rangle^{0}.
\end{aligned}
$$

Að lokum innra vægið $\vec{M}(x)$ sem er fengið á svipaðan hátt og í lið 2(b) með krossmargfeldi stefnuvigurs úr punkti $\vec{p} = [x, 0, 0]^T$ í átakspunkt krafts, og innri skerkraftanna en staka undirstöðuvægið er tekið með öfugu formerki fyrir innra vægi (hækkum veldi á Macauley falli um 1). Hér sleppum við Macauley föllum á stefnuvigrana og látum þau bara á kraftana.

Innra vægi $\vec{M}(x)$ er því
$$
\begin{aligned}
    \vec{M}(x)
    =
    &- \vec{M}_A \langle x \rangle^{0} \\
    &- (\vec{p}_A - \vec{p}) \times \vec{R}_A \langle x \rangle^{0} \\
    &- (\vec{p}_E - \vec{p}) \times \vec{F}_E \langle x - x_E \rangle^{0} \\
    &- (\vec{p}_C - \vec{p}) \times \vec{F}_C \langle x - x_C \rangle^{0}.
\end{aligned}
$$

```python
#| echo: true
#| output: true

# teiknum sker og vægisferla fyrir alla ása
import matplotlib.pyplot as plt

# skilgreinum x gildi fyrir hvern ás
x_vals = np.linspace(0, p_D[0]*1000, 1000) # x gildi í cm

# reiknum V(x) og M(x) fyrir hvern ás
def V_fun(x, R_A, M_A): 
    V = (
        - M_A * mac(x, p_A[0], -1)
        - R_A * mac(x, p_A[0], 0)
        - F_E * mac(x, p_E[0], 0)
        - F_C * mac(x, p_C[0], 0)
    )
    return V

def M_fun(x, R_A, M_A):
    p = np.array([x, 0, 0])

    M = (
        - M_A * mac(x, p_A[0], 0)
        - np.cross(p_A - p, R_A * mac(x, p_A[0], 0))
        - np.cross(p_E - p, F_E * mac(x, p_E[0], 0))
        - np.cross(p_C - p, F_C * mac(x, p_C[0], 0))
    )
    return M

V_vals = np.array([V_fun(x/1000, R_A, M_A) for x in x_vals]) # setjum inn cm sem m
M_vals = np.array([M_fun(x/1000, R_A, M_A) for x in x_vals]) # setjum inn cm sem m
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x_vals, V_vals[:, 0], label='V_x(x)')
plt.plot(x_vals, V_vals[:, 1], label='V_y(x)')
plt.plot(x_vals, V_vals[:, 2], label='V_z(x)')
plt.title('Skerkraftsferlar V(x) í x, y og z stefnur')
plt.xlabel('x (cm)')
plt.ylabel('Skerkraftur (N)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim([0, p_D[0]*1000 + 10])
plt.minorticks_on()
plt.legend()
plt.grid()
plt.subplot(2, 1, 2)
plt.plot(x_vals, M_vals[:, 0], label='M_x(x)')
plt.plot(x_vals, M_vals[:, 1], label='M_y(x)')
plt.plot(x_vals, M_vals[:, 2], label='M_z(x)')
plt.title('Beygjuvægi M(x) um x, y og z ás')
plt.xlabel('x (cm)')
plt.ylabel('Beygjuvægi (Nm)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim([0, p_D[0]*1000 + 10])
plt.minorticks_on()
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()  
```

### (c) Staðsetning krítískra spennu
Skoðum gildin á $\vec{M}(x_A)$ þ.e. í punkti A. Það er jákvætt vægi um alla ása. Athugum að þar sem öxull liggur á $x$ ás er $M_x$ vinduvægi (e. torque). Hin vægin, um $y$ og $z$ ás eru beygjuvægi eins og í dæmi 2.

Í punkti A eru því samsett álag af skerspennu vegna vinduvægis $M_x(x_A)$ og normalspennu vegna samett vægis $M_y(x_A)$ og $M_z(x_A)$.

Skerspennan er hæst á yfirborði öxuls í fjarlægð $\rho = r = d/2$ frá miðju og er reiknuð með jöfnunni fyrir skerspennu
$$
    \tau_r
    =
    \frac{M_x d}{2J_x},
$$
þar sem $J_x$ er pólhverftregða um $x$ ás.

Almennt séð ættum við að reikna krossmargfeldi vægisins $\vec{M}(x_A)$ og geislavigurs $\vec{r}_{yz}(\rho, \theta)$ sem liggur $yz$ planinu með upphafspunkt í flatarmiðju $yz$ þversniðs bitans. Lengd vigursins er geislinn $\rho$ og stefnan ákvarðast með horninu $\theta$ sem reiknast jákvætt fyrir snúning frá $+z$ til $+y$ ás. Gefur þetta skerspennuvigurinn $\vec{\tau}_r(x, y)$ í $yz$ þversniðinu með $x$ ás sem normalvigur
$$
    \vec{\tau}_{r}(y, z)
    =
    \left( diag(\vec{J})^{-1} \vec{M}(x_A) \right) \times \vec{r}_{yz}(\rho, \theta).
$$
Þar sem $diag(\vec{J})$ er andhverfa fylkisins sem hefur pólhverftregðurnar $J_x$, $J_y$ og $J_z$ á hornalínunni. Við þurfum bara $J_x$ hér svo við getum sett $J_y=J_z=\infty$ fyrir útreikninga því hún er einungis skilgreind um $x$ ás fyrir $yz$ þversnið. Við getum umritað geislavigurinn $\vec{r}_{yz}$ á kartesískt form sem fall af $y$ og $z$ hnitum í þversniðinu sem
$$
\begin{aligned}
    \vec{r}_{yz}(\rho, \theta)
    &=
    \begin{bmatrix}
        0 \\
        \rho\sin\theta \\
        \rho\cos\theta
    \end{bmatrix}
    &=
    \begin{bmatrix}
        0 \\
        y \\
        z
    \end{bmatrix}.
\end{aligned}
$$
Athugið að yfirborð öxulsins er gefið með $\rho = d/2$ svo krítískt gildi geislavigrinum er $\vec{r}_{yz}(d/2, \theta)=[0, (d/2)\sin\theta, (d/2)\cos\theta]^T$ óháð $\theta$. Við getum einnig notað $(y,z)$ hnit punkts í þvernsiðinu ef þau eru þekkt frekar en hornið $\theta$.

Fyrir beygjuspennurnar vegna vægis um $y$ og $z$ ása getum við fylgt sömu forskrift og í dæmi 2. Núna er þversniðið hringlaga með tvæggja ása samhverfu ($I_y=I_z$). Hornið $\phi$ milli núllássins og $+z$ ássins ákvarðast því eingöngu af gildum vægjanna $M_y(x_A)$ og $M_z(x_A)$ með
$$
    \tan\phi
    =
    \frac{M_y(x_A)}{M_z(x_A)},
$$
þar sem $\phi$ er jákvætt frá $+z$ ás að $+y$ ás.

Gildi normalspenna er sem fyrr (sjá dæmi 2(d))
$$
    \sigma_p
    =
    \left(
    diag(\vec{I})^{-1}     
    \vec{M}(x_A)
    \right)
    \times
    p_n \vec{n}.
$$

Þar sem normalvigurinn á núllás er
$$
    \vec{n}
    =
    \begin{bmatrix}
        0 \\
        \cos\phi \\
        -\sin\phi
    \end{bmatrix},
$$
og ofanvarpið á normalinn í punkti $(y, z)$ er
$$
    p_n
    =
    y \cos\phi
    - z \sin\phi.
$$
Þeir punktar á hringlaga þvernsniðinu sem eru lengst frá núllásnum eru í $\pm 90^{\circ}$ frá núllásnum, eða undir horni $\theta = \phi + 90^{\circ}$ og $\theta = \phi - 90^{\circ}$.

Að lokum bætum við svo ið normalspennunni vegna áskrafts í $x$ stefnu sem myndar heildarnormalspennuna
$$
    \sigma_{x, tot}
    =
    \frac{-R_{Ax}}{A} \hat{i}
    + 
    \left(
    diag(\vec{I})^{-1}     
    \vec{M}(x_A)
    \right)
    \times
    p_n \vec{n}.
$$

```python
#| echo: true
#| output: true

# reiknum horn theta fyrir krítíska punkta
theta_1 = phi + np.pi/2
theta_2 = phi - np.pi/2
print(f"Horn í punkti 1 theta_1 = {np.degrees(theta_1):.2f}° frá +z ás")
print(f"Horn í punkti 2 theta_2 = {np.degrees(theta_2):.2f}° frá +z ás")

# þvermál öxuls
d = 12/1000 # m

# x og y hnit þessara punkta eru
p_1 = np.array([
    0,
    (d/2) * np.sin(theta_1),
    (d/2) * np.cos(theta_1)
]) 
p_2 = np.array([
    0,
    (d/2) * np.sin(theta_2),
    (d/2) * np.cos(theta_2)
])
print(f"\nHnit punkts 1 eru p_1 = (y, z) = {p_1[1:]*1e3} mm frá miðju")
print(f"Hnit punkts 2 eru p_2 = (y, z) = {p_2[1:]*1e3} mm frá miðju")

# þversniðsflatarmál A
A = (math.pi * d**2) / 4 # m^2
print(f"\nÞversniðsflatarmál A = {A*1e6:.3f} mm^2")

# póltregður J
J_x = (math.pi * d**4) / 32 # m^4
J_y = np.inf
J_z = np.inf
diag_J = np.diag([J_x, J_y, J_z])

# prentum póltregðu J_x
print(f"\nPóltregða J_x = {J_x:.3e} m^4")

# flatartregður I
I_y = (math.pi * d**4) / 64 # m^4
I_z = (math.pi * d**4) / 64 # m^4
I_x = np.inf
diag_I = np.diag([I_x, I_y, I_z])

# prentum hverfisflatartregður I_y og I_z
print(f"\nFlatartregða I_y = {I_y:.3e} m^4")
print(f"Flatartregða I_z = {I_z:.3e} m^4")

tau_1 = np.cross(np.linalg.inv(diag_J) @ M_fun(p_A[0], R_A, M_A), p_1)
tau_2 = np.cross(np.linalg.inv(diag_J) @ M_fun(p_A[0], R_A, M_A), p_2)
print(f"\nSkerspenna í punkti 1 er tau_1 = {tau_1/1e6} MPa")
print(f"Norm skerspennu í punkti 1 er |tau_1| = {np.linalg.norm(tau_1)/1e6} MPa")
print(f"\nSkerspenna í punkti 2 er tau_2 = {tau_2/1e6} MPa")
print(f"Norm skerspennu í punkti 2 er |tau_2| = {np.linalg.norm(tau_2)/1e6} MPa")

# reiknum normalvigur í krítíska punkta
norm_vec_1 = np.array([0, np.cos(theta_1), -np.sin(theta_1)])
norm_vec_2 = np.array([0, np.cos(theta_2), -np.sin(theta_2)])

# reiknum normalspennur í punktum 1 og 2
sigma_1 = np.cross(np.linalg.inv(diag_I) @ M_fun(p_A[0], R_A, M_A), p_1)
sigma_1[0] += -R_A[0]/A # bætum við hreinni normalspennu
sigma_2 = np.cross(np.linalg.inv(diag_I) @ M_fun(p_A[0], R_A, M_A), p_2)
sigma_2[0] += -R_A[0]/A # bætum við hreinni normalspennu
print(f"\nNormalspenna í punkti 1 er sigma_1 = {sigma_1/1e6} MPa")
print(f"Normalspenna í punkti 2 er sigma_2 = {sigma_2/1e6} MPa")

# reiknum skerspennu vegna beygju í miðju þversniði (y,z)=(0,0)
yz_vec = np.array([0, 1, 1])
tau_bending = 4 * np.linalg.norm(np.diag(yz_vec) @ V_fun(p_A[0], R_A, M_A)) / (3 * A)
print(f"\nSkerspenna vegna beygju í miðju þversniði er tau_bending = {tau_bending/1e6:.3f} MPa")
```

Þar með erum við með þrýstispennu $\sigma_1 = -107 \text{ MPa}$ í punkti 1 og togspennu $\sigma_2 = 112 \text{ MPa}$ í punkti 2. Hrein normalspenna vegna ásskrafts hefur hverfandi áhrif á heildarnormal spennur. Í báðum punktum er skerspenna sem er snertill við yfirborð öxuls og rangsælis stefnu í báðum punktum með stærð $\tau_{r, max} = 35.4 \text{ MPa}$. Tökum eftir að hæsta skerspennan í miðju þversniði er $2.36 \text{ MPa}$ sem er mun lægra gildi en skerspennan við yfirborð öxulsins, svo flatarmiðjan er ekki krítísk staðsetning.

### (d) Höfuðspennur og horn höfuðplana
Normalspennurnar í krítískum punktum $1$ og $2$ eru báðar eftir $x$ ás. Skerspennurnar eru í $yz$ planinu. Er því ekki um hreint tvívítt spennuástand að ræða því við erum með skerspennur $\tau_{xy}$ og $\tau_{xz}$. Þyrftum því að skoða tvö plön, annars vegar $xy$ og hins vegar $xz$ planið. Getum þó slegið tvær flugur í einu höggi og skoðað spennuelement í hnitakerfi sem hefur verið snúið um $\phi$, köllum það $xy'z'$ (hefur saman $x$ ás en aðra $y$ og $z$ ása en aðalhnitakerfið). Athugið að $+z'$ er samsíða núllás þversniðsins og $y'$ er normall á núllásinn. Spennuelement í $xy'z'$ hnitakerfi í punktum 1 og 2 er í tvívíðu spennuástandi eins og við höfum skoðað í fyrirlestrum, því þá er $\tau_{xy'} = 0$ og $\tau_{y'z'} = \|\vec{\tau}\|$ og $\sigma_{x} = \sigma_{x, tot}$ sem er óbreytt.

Reiknum höfuðspennur í báðum punktum með (ath $\sigma_{y'}=0$)
$$
    \sigma_1, \sigma_2
    =
    \frac{\sigma_{x} + \sigma_{y'}}{2}
    \pm
    \sqrt{
        \left( \frac{\sigma_{x} - \sigma_{y'}}{2} \right)^2
        + \tau_{xz'}^2
    }.
$$
og hæstu skerspennur
$$
    \tau_{max}
    =
    \sqrt{
        \left( \frac{\sigma_{x} - \sigma_{y'}}{2} \right)^2
        + \tau_{xz'}^2
    }.
$$

Finnum einnig horn höfuðplana með
$$
    \tan 2\theta_p
    =
    \frac{2\tau_{xz'}}{\sigma_x - \sigma_{y'}}
    =
    \frac{2\tau_{xz'}}{\sigma_x - \sigma_{y'}}.
$$
Ath að það eru tvær lausnir $\theta_p$ sem hafa $90^{\circ}$ á milli sín, ein fyrir hverja höfuðspennu.

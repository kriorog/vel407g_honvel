# Dæmi 1
Gegnheill öxull með þvermál $d$ verður fyrir bæði jafndreifðri normalspennu $\sigma_n$ vegna áskrafts $P$ sem og beygjuspennu $\sigma_b$ vegna vægis $M$. Hæsta spenna á ytra yfirborði öxuls er $\sigma_{max}=\sigma_n+\sigma_b$.

Áskraftur $P$ hefur meðaltal $\mu_P=779 \text{ kN}$ og staðalfrávik $\hat{\sigma}_P = 72.7\text{ kN}$. Beygjuvægið hefur meðaltal $\mu_M=43.5 \text{ kNm}$ og staðalfrávik $\hat{\sigma}_M=2.53 \text{ kNm}$.

Öxull er úr stáli sem hefur meðalflotstyrk $\bar{S}_y = 553 \text{ MPa}$ og staðalfrávik $\hat{\sigma}_{S_y}=42.7 \text{ MPa}$. 

Gerið ráð fyrir að allar stærðir séu óháðar og normaldreifðar.

Ákvarðið hönnunarstuðul og þvermál öxuls m.v. 99.0\% áreiðanleika (öryggi) gegn flotbjögun.

## Lausn
Normalspennan er $\sigma_n = P/A$ og hæsta beygjuspenna er $\sigma_b = Mr/I = Md/2I$ þar sem $r=d/2$ er geisli öxuls. Þar sem hvor spennan fyrir sig er línulegt fall af normaldreifðri slembibreytu, þ.e. $P$ og $M$, þá eru spennurnar einnig normaldreifðar. 

Normalspennan $\sigma_n$ er því normaldreifð með meðaltal $\mu_{\sigma_n} = \mu_{P}/A$ og staðalfrávik $\hat{\sigma}_{\sigma_n} = \hat{\sigma}_{P}/A$. Hæsta beygjuspennan er einnig normaldreifð með meðaltal $\mu_{\sigma_b}=\mu_{M}d/2I$ og staðalfrávik $\hat{\sigma}_{\sigma_b}=\hat{\sigma}_M d/2I$.

Einnig er hæsta spenna $\sigma_{max}$ normaldreifð slembibreyta því hún er summa $\sigma_n$ og $\sigma_b$. Hæsta spennan hefur því meðaltal 

$$
\begin{aligned}
    \mu_{\sigma_{max}}  &= \mu_{\sigma_n} + \mu_{\sigma_b}, \\
                        &= \frac{\mu_P}{A} + \frac{\mu_M d}{2I},
\end{aligned}
$$
og staðalfrávik

$$
\begin{aligned}
    \hat{\sigma}_{\sigma_{max}} &= \sqrt{\left(\frac{\hat{\sigma}_P}{A}\right)^2 + \left(\frac{\hat{\sigma}_Md}{2I}\right)^2}.
\end{aligned}
$$

Fyrir hringlaga gegnheilt þversnið með þvermál $d$ er $A=\frac{\pi d^2}{4}$ og $I=\frac{\pi d^4}{64}$. Getum því ritað $I = \frac{d^2}{16}A$. Þar með má umrita meðaltalið og staðalfrávikið

$$
    \mu_{\sigma_{max}} = \frac{1}{A}\left( \mu_P + \frac{8\mu_M}{d}\right),
$$
og
$$
    \hat{\sigma}_{\sigma_{max}} = \frac{1}{A}\sqrt{\hat{\sigma}_P^2 + \left(\frac{8\hat{\sigma}_M}{d}\right)^2}.
$$

:::{.callout-note title="Meðhöndlun á normaldreifðum slembibreytum"}
Til að staðfesta þessar niðurstöður[^1], þá þarf að athuga að líkindadreifingar $\sigma_n$ og $\sigma_b$ eru fengnar með því að skala normaldreifingar á óháðu slembibreytunum $P$ og $M$ með fasta. Ef slembibreyta $X$ er normaldreifð, þá er breytan $aX$ fyrir eitthvað $a \in \mathbb{R}$ einnig normaldreifð með meðaltal $\mu_{aX} = a\mu_{X}$ og staðalfrávik $\hat{\sigma}_{aX}=|a|X$. Ef $a \geq 0$ þá má sleppa tölugildinu $|\cdot|$ í staðalfrávikinu.

Ef $Y$ er önnur normaldreifð slembibreyta óháð $X$, þá er $Z=X+Y$ einnig normaldreifð slembibreyta með meðaltal $\mu_z = \mu_x + \mu_z$ og staðalfrávik $\hat{\sigma}_Z=\sqrt{\hat{\sigma}_{\sigma_n}^2+ \hat{\sigma}_{\sigma_b}^2}$.

[^1]: Sjá til dæmis Ross *Introduction to Probability and Statistics* kafla 4.5 og 4.6 úr STÆ203G "Lík og töl".
:::

Þar með höfum við sett fram líkindadreifingu normaldreifðu slembibreytunnar $\sigma_{max}$ sem fall af þeim upplýsingum sem við höfum gefin í verkefninu, þ.e. meðaltöl og staðalfrávik álagsins $P$ og beygjuvægisins $M$.

Jafna (1-12) í Shigley kafla 1 tengir saman áreiðanleika (öryggi) í gegnum samsvarandi $z$ gildi staðlaðrar normaldreifingar og viðeigandi hönnunarstuðuls $n_d$. 

$$
    n_d = \frac{1 \pm \sqrt{1 - (1-z^2C_S^2)(1-z^2C_{\sigma_{max}}^2)}}{1-z^2C_S^2}
$$

þar sem $C_S=\hat{\sigma}_S/\mu_S$ og $C_{\sigma_{max}}=\hat{\sigma}_{\sigma_{max}}/\mu_{\sigma_{max}}$ eru hjálparbreytur fyrir styrk og spennu. Ath að $1/A$ styttist út fyrir $C_{\sigma_{max}}$.

Fyrir $99.0\%$ öryggi, þá getum við leitað uppi $z$ gildi í töflum fyrir stofnfall líkindadreififallsins þ.e. $\Phi(z) = 0.990$ og jafnvel brúað á milli gildi ef þess þarf. Einnig er hægt að nota Python og Scipy pakkann og nota andhverfu $\Phi$ til að gefa okkur $z$ gildi fyrir ákveðið öryggi.

Athugum að $C_S$ er fasti en $C_{\sigma_{max}}$ er fall af $d$. Við getum því ritað hönnunarstuðulinn sem fall af $z$ og $d$ sem 
$$ n_d = n_d(z, d) = f(z, d),$$

þar sem $f(z,d)$ er hægri hliðin í jöfnu $n_d$ fyrir ofan. Við getum því ekki leyst fyrir hönnunarstuðlinn $n_d$ sem fall af áreiðanleikanum eingöngu (með $z$) heldur þarf einnig að taka tillit til þvermáls öxulsins, því líkindadreifing beygjuspennunnar $\sigma_b$ er fall af þvermálinu. Hönnunarstuðullinn þarf alltaf að vera $\geq 1$ svo fyrir gefinn hönnunarstuðull $n_d = a \geq 1$ má leysa fyrir þvermál $d$ sem núllstöð  ólínulegu jöfnunnar
$$ a - f(z_0, d) = 0, $$
Þvermálið $d$ fyrir $a=1$ er lægsta fræðilega þvermálið sem gengur upp. Þar sem $f(z, d)$ er ekki eifnöld margliða, þá getum við leyst fyrir núllstöðina með *fsolve* fallinu í Python pakkanum *scipy.optimize*.

```python
#| echo: true
#| output: true

from scipy.stats import norm
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt


# finnum z gildi fyrir 99.0% áreiðanleika (öryggi)
r = 0.990 # 99% reliability
z = norm.ppf(1-r) # inverse \Phi(z)=r i.e. z=\Phi^-1(r)
print(f"z gildi fyrir gefið örygg {r:.3f} er {z:.3f}")

# gefið meðaltal og staðalfrávik á krafti P.
mu_p = 779e3 # N
stdev_p = 72.7e3 # N

# gefið meðaltal og staðalfrávik á beygjuvægi M.
mu_m = 43.5e3 # Nm
stdev_m = 2.53e3 # Nm

# gefið meðaltal og staðalfrávik á styrk 
mu_S = 553e6 # MPa
stdev_S = 42.7e6 # MPa
C_S = stdev_S/mu_S

# skilgreinum lambda fall fyrir meðaltal hæstu heildarspennu 
mu_max = lambda d: mu_p + 8*mu_m/d

# skilgreinum lambda fall fyrir staðalfrávik hæstu heildarspennu 
stdev_max = lambda d: np.sqrt(stdev_p**2 + (8*stdev_m/d)**2)


# skilgreinum fallið f(d) fyrir gefið z og a
# höfum d í mm svo við setjum inn d/1000
a = 1
def f(d):
    num = 1 + np.sqrt(
                1
                - (1 - (z*C_S)**2)
                * (1 - (z*stdev_max(d/1000)/mu_max(d/1000))**2)
            )

    den = 1 - (z*C_S)**2

    fval = num/den

    return fval

d_vals = np.linspace(1, 1000, 1000)
f_vals = f(d_vals)

plt.figure(figsize=(10, 6))
plt.plot(d_vals, f_vals, label='f(d)')
plt.xlabel('d')
plt.ylabel('f(d)')
plt.title('Function Plot')
plt.grid(True)
plt.legend()
plt.ylim([0,1.5])
plt.show()

# skilgreinum rótarverkefnið a - f(z, d) = 0 fyrir gefið z og a
d0 = 100 # upphafsgisk á d í mm
#roots = fsolve(f, d0)
#print(roots)


A=np.pi*(105/1000)**2/4
print(mu_max(105)/A/1e6)
print(stdev_max(105)/A/1e6)
print(f(106))

plt.figure("Figure 2")
plt.plot(d_vals, stdev_max(d_vals)/mu_max(d_vals), label='C_sigma')
plt.xlabel('d [mm]')
plt.ylabel('C_sigma')

n_d = 1.256
S_max=mu_S/n_d
print(S_max)
d = np.roots([np.pi*S_max/4, 0, -mu_p, - 8*mu_m])
print(d*1000)
d_new = np.real(d[0])
A = lambda d: np.pi*(d)**2/4
z_d = ((mu_p+8*mu_m/d_new)/A(d_new)-mu_S)/np.sqrt(stdev_S**2 + (stdev_p**2 + (8*stdev_m/d_new)**2)/A(d_new)**2)
print(z_d)
1-norm.cdf(z_d)
```

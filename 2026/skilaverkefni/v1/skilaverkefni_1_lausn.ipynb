# Dæmi 1
Gegnheill öxull með þvermál $d$ verður fyrir bæði jafndreifðri normalspennu $\sigma_n$ vegna áskrafts $P$ sem og beygjuspennu $\sigma_b$ vegna vægis $M$. Hæsta spenna á ytra yfirborði öxuls er $\sigma_{max}=\sigma_n+\sigma_b$.

Áskraftur $P$ hefur meðaltal $\mu_P=779 \text{ kN}$ og staðalfrávik $\hat{\sigma}_P = 72.7\text{ kN}$. Beygjuvægið hefur meðaltal $\mu_M=43.5 \text{ kNm}$ og staðalfrávik $\hat{\sigma}_M=2.53 \text{ kNm}$.

Öxull er úr stáli sem hefur meðalflotstyrk $\bar{S}_y = 553 \text{ MPa}$ og staðalfrávik $\hat{\sigma}_{S_y}=42.7 \text{ MPa}$. 

Gerið ráð fyrir að allar stærðir séu óháðar og normaldreifðar.

Ákvarðið hönnunarstuðul og þvermál öxuls m.v. 99.0\% áreiðanleika (öryggi) gegn flotbjögun.

## Lausn
Normalspennan er $\sigma_n = P/A$ og hæsta beygjuspenna er $\sigma_b = Mr/I = Md/2I$ þar sem $r=d/2$ er geisli öxuls. Þar sem hvor spennan fyrir sig er línulegt fall af normaldreifðri slembibreytu, þ.e. $P$ og $M$, þá eru spennurnar einnig normaldreifðar. 

Normalspennan $\sigma_n$ er því normaldreifð með meðaltal $\mu_{\sigma_n} = \mu_{P}/A$ og staðalfrávik $\hat{\sigma}_{\sigma_n} = \hat{\sigma}_{P}/A$. Hæsta beygjuspennan er einnig normaldreifð með meðaltal $\mu_{\sigma_b}=\mu_{M}d/2I$ og staðalfrávik $\hat{\sigma}_{\sigma_b}=\hat{\sigma}_M d/2I$.

Einnig er hæsta spenna $\sigma_{max}$ normaldreifð slembibreyta því hún er summa $\sigma_n$ og $\sigma_b$. Hæsta spennan hefur því meðaltal 

$$
\begin{aligned}
    \mu_{\sigma_{max}}  &= \mu_{\sigma_n} + \mu_{\sigma_b}, \\
                        &= \frac{\mu_P}{A} + \frac{\mu_M d}{2I},
\end{aligned}
$$
og staðalfrávik

$$
\begin{aligned}
    \hat{\sigma}_{\sigma_{max}} &= \sqrt{\left(\frac{\hat{\sigma}_P}{A}\right)^2 + \left(\frac{\hat{\sigma}_Md}{2I}\right)^2}.
\end{aligned}
$$

Fyrir hringlaga gegnheilt þversnið með þvermál $d$ er $A=\frac{\pi d^2}{4}$ og $I=\frac{\pi d^4}{64}$. Getum því ritað $I = \frac{d^2}{16}A$. Þar með má umrita meðaltalið og staðalfrávikið

$$
    \mu_{\sigma_{max}} = \frac{1}{A}\left( \mu_P + \frac{8\mu_M}{d}\right),
$$
og
$$
    \hat{\sigma}_{\sigma_{max}} = \frac{1}{A}\sqrt{\hat{\sigma}_P^2 + \left(\frac{8\hat{\sigma}_M}{d}\right)^2}.
$$

:::{.callout-note title="Meðhöndlun á normaldreifðum slembibreytum"}
Til að staðfesta þessar niðurstöður[^1], þá þarf að athuga að líkindadreifingar $\sigma_n$ og $\sigma_b$ eru fengnar með því að skala normaldreifingar á óháðu slembibreytunum $P$ og $M$ með fasta. Ef slembibreyta $X$ er normaldreifð, þá er breytan $aX$ fyrir eitthvað $a \in \mathbb{R}$ einnig normaldreifð með meðaltal $\mu_{aX} = a\mu_{X}$ og staðalfrávik $\hat{\sigma}_{aX}=|a|X$. Ef $a \geq 0$ þá má sleppa tölugildinu $|\cdot|$ í staðalfrávikinu.

Ef $Y$ er önnur normaldreifð slembibreyta óháð $X$, þá er $Z=X+Y$ einnig normaldreifð slembibreyta með meðaltal $\mu_z = \mu_x + \mu_z$ og staðalfrávik $\hat{\sigma}_Z=\sqrt{\hat{\sigma}_{\sigma_n}^2+ \hat{\sigma}_{\sigma_b}^2}$.

[^1]: Sjá til dæmis Ross *Introduction to Probability and Statistics* kafla 4.5 og 4.6 úr STÆ203G "Lík og töl".
:::

Þar með höfum við sett fram líkindadreifingu normaldreifðu slembibreytunnar $\sigma_{max}$ sem fall af þeim upplýsingum sem við höfum gefin í verkefninu, þ.e. meðaltöl og staðalfrávik álagsins $P$ og beygjuvægisins $M$.

Jafna (1-12) í Shigley kafla 1 tengir saman áreiðanleika (öryggi) í gegnum samsvarandi $z$ gildi staðlaðrar normaldreifingar og viðeigandi hönnunarstuðuls $n_d$. 

$$
    n_d = \frac{1 \pm \sqrt{1 - (1-z^2C_S^2)(1-z^2C_{\sigma_{max}}^2)}}{1-z^2C_S^2}
$$

þar sem $C_S=\hat{\sigma}_S/\mu_S$ og $C_{\sigma_{max}}=\hat{\sigma}_{\sigma_{max}}/\mu_{\sigma_{max}}$ eru hjálparbreytur fyrir styrk og spennu. Ath að $1/A$ styttist út fyrir $C_{\sigma_{max}}$.

Fyrir $99.0\%$ öryggi, þá getum við leitað uppi $z$ gildi í töflum fyrir stofnfall líkindadreififallsins þ.e. $\Phi(z) = 0.990$ og jafnvel brúað á milli gildi ef þess þarf. Einnig er hægt að nota Python og Scipy pakkann og nota andhverfu $\Phi$ til að gefa okkur $z$ gildi fyrir ákveðið öryggi.

Athugum að $C_S$ er fasti en $C_{\sigma_{max}}$ er fall af $d$. Við getum því ritað hönnunarstuðulinn sem fall af $z$ og $d$ sem 
$$ n_d = n_d(z, d) = f(z, d),$$

þar sem $f(z,d)$ er hægri hliðin í jöfnu $n_d$ fyrir ofan. Við getum því ekki leyst fyrir hönnunarstuðlinn $n_d$ sem fall af áreiðanleikanum eingöngu (með $z$) heldur þarf einnig að taka tillit til þvermáls öxulsins, því líkindadreifing beygjuspennunnar $\sigma_b$ er fall af þvermálinu. Hönnunarstuðullinn þarf alltaf að vera $\geq 1$ svo fyrir gefinn hönnunarstuðull $n_d = a \geq 1$ má leysa fyrir þvermál $d$ sem núllstöð  ólínulegu jöfnunnar
$$ a - f(z_0, d) = 0, $$
Þvermálið $d$ fyrir $a=1$ er lægsta fræðilega þvermálið sem gengur upp. Þar sem $f(z, d)$ er ekki eifnöld margliða, þá getum við leyst fyrir núllstöðina með *fsolve* fallinu í Python pakkanum *scipy.optimize*.

```python
#| echo: true
#| output: true

from scipy.stats import norm
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt


# finnum z gildi fyrir 99.0% áreiðanleika (öryggi)
r = 0.990 # 99% reliability
z = norm.ppf(1-r) # inverse \Phi(z)=r i.e. z=\Phi^-1(r)
print(f"z gildi fyrir gefið örygg {r:.3f} er {z:.3f}")

# gefið meðaltal og staðalfrávik á krafti P.
mu_p = 779e3 # N
stdev_p = 72.7e3 # N

# gefið meðaltal og staðalfrávik á beygjuvægi M.
mu_m = 43.5e3 # Nm
stdev_m = 2.53e3 # Nm

# gefið meðaltal og staðalfrávik á styrk 
mu_S = 553e6 # MPa
stdev_S = 42.7e6 # MPa
C_S = stdev_S/mu_S

# skilgreinum lambda fall fyrir meðaltal hæstu heildarspennu 
mu_max = lambda d: mu_p + 8*mu_m/d

# skilgreinum lambda fall fyrir staðalfrávik hæstu heildarspennu 
stdev_max = lambda d: np.sqrt(stdev_p**2 + (8*stdev_m/d)**2)


# skilgreinum fallið f(d) fyrir gefið z og a
# höfum d í mm svo við setjum inn d/1000
a = 1
def f(d):
    num = 1 + np.sqrt(
                1
                - (1 - (z*C_S)**2)
                * (1 - (z*stdev_max(d/1000)/mu_max(d/1000))**2)
            )

    den = 1 - (z*C_S)**2

    fval = num/den

    return fval

d_vals = np.linspace(1, 1000, 1000)
f_vals = f(d_vals)

plt.figure(figsize=(10, 6))
plt.plot(d_vals, f_vals, label='f(d)')
plt.xlabel('d')
plt.ylabel('f(d)')
plt.title('Function Plot')
plt.grid(True)
plt.legend()
plt.ylim([0,1.5])
plt.show()

# skilgreinum rótarverkefnið a - f(z, d) = 0 fyrir gefið z og a
d0 = 100 # upphafsgisk á d í mm
#roots = fsolve(f, d0)
#print(roots)


A=np.pi*(105/1000)**2/4
print(mu_max(105)/A/1e6)
print(stdev_max(105)/A/1e6)
print(f(106))

plt.figure("Figure 2")
plt.plot(d_vals, stdev_max(d_vals)/mu_max(d_vals), label='C_sigma')
plt.xlabel('d [mm]')
plt.ylabel('C_sigma')

n_d = 1.256
S_max=mu_S/n_d
print(S_max)
d = np.roots([np.pi*S_max/4, 0, -mu_p, - 8*mu_m])
print(d*1000)
d_new = np.real(d[0])
A = lambda d: np.pi*(d)**2/4
z_d = ((mu_p+8*mu_m/d_new)/A(d_new)-mu_S)/np.sqrt(stdev_S**2 + (stdev_p**2 + (8*stdev_m/d_new)**2)/A(d_new)**2)
print(z_d)
1-norm.cdf(z_d)
```

```python
from scipy.stats import norm
import numpy as np

# determine z score for given reliability
r = 0.990 # 99% reliability
z = norm.ppf(1-r) # inverse \Phi(z)=r i.e. z=\Phi^-1(r)
print(f"z gildi fyrir gefið örygg {r:.3f} er {z:.3f}")

# given values for mean and stdev of load P
mu_p = 779 # kN
stdev_p = 72.7 # kN

# given values for mean and stdev of moment M
mu_m = 43.5 # kNm
stdev_m = 2.53 # kNm

# mean and stdev of normal maximum stress
mu_stress = mu_p + mu_m
stdev_stress = np.sqrt(stdev_p**2 + stdev_m**2)

# distribution of strengs
mu_strength = 553 # MPa
stdev_strength = 42.7 # MPa

# help variables C for each random variable
C_stress = stdev_stress/mu_stress
C_strength = stdev_strength/mu_strength

# calculate the required design factor
nd = 1 + np.sqrt(1 - (1 - z**2*C_strength**2)*(1 - z**2*C_stress**2))/(1 - z**2*C_strength**2)
print(nd)

# maximum allowable stress
stress_max = mu_strength/nd
print(stress_max)

# solve the cubic equation - bring all coeffs to base units (Pa, N, Nm)
cubic_coeff = [(stress_max*1e6)*np.pi, 0, -4*(mu_p*1000), -32*(mu_m*1000)]
d_roots = np.roots(cubic_coeff)
print(d_roots)

# only one real solution, which is the d in meters, multiply with 1000 for mm
print(np.real(d_roots[0])*1000)
```

## Lausn
Í þessu verkefni liggur beinast við að skoða $xy$ og $xz$ plönin í sitthvoru lagi, setja upp Macauley föllin, finna viðeigandi undirstöðukrafta, sem og innri skerkrafta og vægi. Er það alveg góð og gild nálgun. Hins vegar, þar sem þetta verkefni er þrívítt með $xyz$ stefnur, þá er áskorun að tryggja rétt formerki á kröftum og vægjum skv. því hnitakerfi sem er skilgreint á myndinni með því eingöngu "að sjá það".

Eina leiðin til að tryggja rétt formerki er að vinna með vigra fyrir alla krafta og vægi í þrívíðu hnitakerfi. Þetta getur flækt framsetningu töluvert og reynst erfiðara að reikna í höndunum, sérstaklega þegar kemur að vægjum þar sem krossmargfeldi þarf að nota til að finna réttan vægisvigur með réttri stefnu.

Vil ég því nýta tækifærið með þessu dæmi og kynna Macauley föll með vigra í þrívíðu hnitakerfi. Þetta getur verið gagnlegt fyrir þá sem vilja læra að vinna með vigra í þessum verkefnum og tryggja rétt formerki á kröftum og vægjum - til hliðar við þá almennu aðferð sem kynnt var í Burðarþolsfræði (VÉL202G).

::: {.callout-note title="Macauley föll á þrívíða vigra"}
Macauley föllin sem eru skilgreind í Töflu 3-1 í Shigley gera ráð fyrir skalargildum á staðsetningu. Til þess að geta beitt þessum föllum á þrívíða vigra þá getum við ákveðið að aðlaga skilgreininguna til þess að taka inn stefnuvigur $\vec{p}$ og bera hann saman við punkt $\vec{a}$ í rúminu
$$
    \langle \vec{p} - \vec{a} \rangle^n = 
    \langle p_x - a_x \rangle^0
    \langle p_y - a_y \rangle^0
    \langle p_z - a_z \rangle^0
    \begin{bmatrix}
        (p_x - a_x)^n \\
        (p_y - a_y)^n \\
        (p_z - a_z)^n
    \end{bmatrix},
$$
sem mun gefa núllvigurinn nema ef öll hnit stefnuvigursins $\vec{p}$ eru jöfn eða stærri en samsvarandi hnits í punktinum $\vec{a}$.

Við munum einnig þurfa að vinna með krossmargfeldi Macauley vigurfalls og kraftavigurs $\vec{F}$ til þess að reikna vægisvigur $\vec{M}$

$$
    \vec{M} = \langle \vec{p} - \vec{a} \rangle^n \times \vec{F},
$$
:::

Í þessu verkefni liggur bitinn á $x$ ásnum, svo öll $y$ og $z$ hnit mögulegra punkta $\vec{a}$ eru núll. Sama má segja um mögulega stefnuvigra $\vec{p}$ sem hafa bara $x$ hnit. Þar sem mögulegir punktar eru $\vec{a} = [a_x, 0, 0]^T$ og stefnuvigrar $\vec{p} = [p_x, 0, 0]^T$ eru mögulegir Macauley vigrar sem við þurfum að vinna með

$$
    \langle \vec{p} - \vec{a} \rangle^n =
    \langle p_x - a_x \rangle^0
    (p_x - a_x)^n
    \begin{bmatrix}
        1 \\
        0 \\
        0
    \end{bmatrix}.
$$
Einnig eru allir kraftvigrar í $yz$ plani, þ.e. hafa $x$ hnit núll. Þar með einfaldast möguleg krossmargfeldi í verkefninu töluvert

$$
\begin{aligned}
    \vec{M} = \langle \vec{p} - \vec{a} \rangle^n \times \vec{F} &=
    \langle p_x - a_x \rangle^0
    \begin{vmatrix}
    \hat{i} & \hat{j} & \hat{k} \\
    \langle p_x - a_x \rangle^n & 0 & 0 \\
    0 & F_y & F_z
    \end{vmatrix}\\
    &= 
    \langle p_x - a_x \rangle^0
    \begin{bmatrix}
        0 \\
        -\langle p_x - a_x \rangle^n F_z \\
        \langle p_x - a_x \rangle^n F_y
    \end{bmatrix} \\
    &= 
    \langle p_x - a_x \rangle^0
    (p_x - a_x)^n 
    \begin{bmatrix}
        0 \\
        -F_z \\
        F_y
    \end{bmatrix}.
\end{aligned}
$$
Tökum eftir að jákvæður kraftur í $z$ stefnu veldur neikvæðu vægi um $y$ ásinn, en jákvæður kraftur um $y$ ás veldur jákvæðu vægi um $z$ ásinn.


### (a) Macauley föll
Álagið $q(x)$ er 
$$
    q(x) = \vec{R_1} \langle x \rangle^{-1} + \vec{P}_A \langle x-x_A \rangle^{-1} + \vec{w}_{BC}\langle x-x_B\rangle^0 - \vec{w}_{BC}\langle x-x_C\rangle^0 + \vec{R}_2\langle x - x_C\rangle^{-1}.
$$
Ath að við höfum ekki sett inn stefnur vigrana því stefnan er skilgreind með formerkjum hnita í hverjum vigri. Miðað við verkefnið þá er $\vec{R}_1=[0, R_{1y}, R_{1z}]^T$, $\vec{P}_A = [0, 0, P_{Az}]^T$, $\vec{w}_{BC} = [0, -w_{BC}\cos{\theta}, -w_{BC}\sin{\theta}]^T$ og $\vec{R}_2 = [0, R_{2y}, R_{2z}]^T$, þar sem $P_{Az}=1.50 \text{ kN}$, $w_{BC}=2.00 \text{ kN/m}$ og $\theta = 30^\circ$. Einnig eru staðsetningar $x_A=0.500 \text{ m}$, $x_B = 1.50 \text{ m}$ og $x_C = 4.00 \text{ m}$.

Finnum skerkraftinn $V(x)$ með því að heilda $q(x)$ einu sinni og margfalda með -1, því innri skerkraftur vegna jákvæðs ytri krafts $\vec{F}$ er $\vec{V}=-\vec{F}$.
$$
    V(x) = 
    -\vec{R_1} \langle x \rangle^{0} 
    - \vec{P}_A \langle x-x_A \rangle^{0} 
    -\vec{w}_{BC}\langle x-x_B\rangle^{1} 
    + \vec{w}_{BC}\langle x-x_C\rangle^{1} 
    - \vec{R}_2\langle x - x_C\rangle^{0}.
$$

Til þess að finna $M(x)$ þá þurfum við að heilda Macauley föllin úr $V(x)$ einu sinni til viðbótar, en í stað þess að margfalda þau við kraftana, þá þarf að rita stefnur sem vigra og krossmargfalda við krafta.
$$
\begin{aligned}
    M(x) &= 
    \langle \vec{x} \rangle^1 \times \vec{R}_1 \\
    &+ \langle \vec{x} - \vec{x}_A \rangle^ 1 \times \vec{P}_A \\
    &+ \langle \vec{x}-\vec{x}_B\rangle^{2} \times \frac{\vec{w}_{BC}}{2} \\
    &- \langle \vec{x}-\vec{x}_C\rangle^{2} \times \frac{\vec{w}_{BC}}{2} \\
    &+ \langle \vec{x} - \vec{x}_C\rangle^{1} \times \vec{R}_2.
\end{aligned}
$$

Athugið að ofangreind heildun, eftir að hún er framkvæmd, er óákveðin þar sem engin heildismörk eru sett. Hins vegar þurfum við ekki heildunarfasta í hverri heildun, því jaðarskilyrði á $V(x)$ og $M(x)$ í $x=0$ og $x=X_B=3.00 \text{ m}$ eru uppfyllt án þeirra. Sjáum að $V(0)=\vec{R}_1$ og $V(x_C)=\vec{0}$ því $V(x_C)$ er summa allra krafta sem verður að vera núllvigurinn vegna kraftajafnvægsi. Á sama hátt er $M(0)=\vec{0}$ og $M(x_C)=\vec{0}$ því $M(x_C)$ er summa allra vægja um punkt $C$ sem verður einnig að vera núllvigurinn vegna vægjajafnvægis um $y$ og $z$ ása.

### (b) Undirstöðukraftar
Við getum sett upp jöfnuhneppi til að finna undirstöðukraftana með því að skoða $\vec{V}(x)$ og $\vec{M}(x)$ úr a lið. Fyrir kraftajafnvægi reiknum við $\vec{V}(x_C)$ sem

$$
    \vec{V}(x_C) = -\vec{R_1} - \vec{P}_A - \vec{w}_{BC}(x_C - x_B) + \vec{w}_{BC}(x_C - x_C) - \vec{R}_2 = \vec{0}.
$$
Ath að $\vec{w}_{BC}(x_C-x_C)=\vec{0}$. Aðgreinum óþekktar og þekktar stærðir
$$
    \vec{R}_1 + \vec{R}_2 = -\vec{P}_A - \vec{w}_{BC}(x_C - x_B).
$$

Reiknum einnig $\vec{M}(\vec{x}_C)$ 
$$
\begin{aligned}
    M(\vec{x}_C) &= 
    \langle \vec{x}_C \rangle^1 \times \vec{R}_1 \\
    &+ \langle \vec{x}_C - \vec{x}_A \rangle^ 1 \times \vec{P}_A \\
    &+ \langle \vec{x}_C-\vec{x}_B\rangle^{2} \times \frac{\vec{w}_{BC}}{2} \\
    &- \langle \vec{x}_C-\vec{x}_C\rangle^{2} \times \frac{\vec{w}_{BC}}{2} \\
    &+ \langle \vec{x}_C - \vec{x}_C\rangle^{1} \times \vec{R}_2 = \vec{0}.
\end{aligned}
$$

Athugum að $\langle \vec{x}_C - \vec{x}_C \rangle^n = \vec{0}$. Aðgreinum óþekktar og þekktar stærðir

$$
\begin{aligned}
    \langle \vec{x}_C \rangle^1 \times \vec{R}_1 &=
    - \langle \vec{x}_C - \vec{x}_A \rangle^ 1 \times \vec{P}_A \\
    &- \langle \vec{x}_C-\vec{x}_B\rangle^{2} \times \frac{\vec{w}_{BC}}{2}.
\end{aligned}
$$

Miðað við okkar skilgreiningu á krossmargfeldi vigurgilds Macauleyfalls og kraft er
$$
\begin{aligned}
    \langle \vec{x_c} \rangle^1 \times \vec{R}_1 &= 
    \langle x_c \rangle^0
    \begin{bmatrix}
        x_c\\
        0 \\
        0
    \end{bmatrix}
    \times
    \begin{bmatrix}
        0 \\
        R_{1y} \\
        R_{1z} \\
    \end{bmatrix} \\
    &=
    x_c
    \begin{bmatrix}
        0 \\
        -R_{1z} \\
        R_{1y}
    \end{bmatrix}.
\end{aligned}
$$

Þar með höfum við fundið undirstöðukraftinn $\vec{R}_1$ í punkti $O$ sem
$$
\begin{aligned}
    \begin{bmatrix}
        0 \\
        -R_{1z} \\
        R_{1y}
    \end{bmatrix}
    &=
    \frac{1}{x_c}
    \left(
        - \langle \vec{x}_C - \vec{x}_A \rangle^1 \times \vec{P}_A
        - \langle \vec{x}_C-\vec{x}_B\rangle^{2} \times \frac{\vec{w}_{BC}}{2}
    \right),
\end{aligned}
$$
sem við getum leyst með útreikningum einum og sér því öll hægri hliðin inniheldur þekktar stærðir.

Að lokum getum við svo fundið undirstöðukraftinn $\vec{R}_2$ í punkti $C$ sem
$$  
    \vec{R}_2 = -\vec{P}_A - \vec{w}_{BC}(x_C - x_B) - \vec{R}_1.
$$ 

```python
#| echo: true
#| output: true

import numpy as np

# skilgreinum macauley fall
mac = lambda x, a, n: 0 if (n < 0) or (x < a) else (x - a) ** n
    
# skilgreinum gefnar stærðir
P_A = np.array([0, -1500]) # N
w = 2000 # N/m
theta = 30 * np.pi / 180 # radians
x_A = 0.500 # m
x_B = 1.500 # m
x_C = 3.000 # m
w_BC = np.array([-w*np.cos(theta), -w*np.sin(theta)]) # N/m

# skilgreinum V(x) og M(x) sem föll
def V_fun(x,R_1, R_2): 
    V = (
        R_1*mac(x, 0, 0)
        + P_A * mac(x, x_A, 0)
        + w_BC * mac(x, x_B, 1)
        - w_BC * mac(x, x_C, 1)
        + R_2 * mac(x, x_C, 0)
    )
    return V

def M_fun(x, R_1, R_2):
    M = (
        R_1 * mac(x, 0, 1)
        + P_A * mac(x, x_A, 1)
        + w_BC * mac(x, x_B, 2) / 2
        - w_BC * mac(x, x_C, 2) / 2
        + R_2 * mac(x, x_C, 1)
    )
    return M

# reiknum undirstöðukrafta skv V(x) og M(x)
R_1 = -(
    (
        P_A * (x_C - x_A) 
        + 0.5 * w_BC * (x_C - x_B) ** 2
    )
    / x_C
)

R_2 = -P_A - w_BC * (x_C - x_B) - R_1

print(f"Prentum niðurstöður og staðfestum jafnvægi V(x_C)=0 og M(x_C)=0")
print(f"R_1 = [R_1y, R_1z] = {R_1} N")
print(f"R_2 = [R_2y, R_2z] = {R_2} N")
print(f"V_fun(x_C, R_1, R_2) = {V_fun(x_C, R_1, R_2)} N")
print(f"M_fun(x_C, R_1, R_2) = {M_fun(x_C, R_1, R_2)} Nm")
```

Svo við höfum undirstöðukraftana $\vec{R}_1$ og $\vec{R}_2$ með 3 markverða tölustafi
$$
    \vec{R}_1 = 
    \begin{bmatrix}
        R_{1y} \\ 
        R_{1z}
    \end{bmatrix}
    =
    \begin{bmatrix}
        0.650 \\ 
        1.63 
    \end{bmatrix} \text{ kN},
$$
og
$$
    \vec{R}_2 = 
    \begin{bmatrix}
        R_{2y} \\ 
        R_{2z}
    \end{bmatrix}
    =
    \begin{bmatrix}
        1.95 \\ 
        1.38 
    \end{bmatrix} \text{ kN}.
$$  

### (c) Sker- og vægisferlar
Nú getum við teiknað sker- og vægisferla í bæði $xy$ og $xz$ plönum með því að nota Macauley föllin úr a lið. Við getum notað Python og Matplotlib til að teikna ferlana. Hér eru sýndar myndirnar fyrir sker- og vægisferla í bæði $xy$ og $xz$ plönum.

```python
#| echo: true
#| output: true

# teiknum V(x) og M(x) yfir bilið [0, x_C]
x_vals = np.linspace(0, x_C, 1000)
V_vals = np.array([V_fun(x, R_1, R_2) for x in x_vals])
M_vals = np.array([M_fun(x, R_1, R_2) for x in x_vals])

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x_vals, V_vals[:, 0], label='V_y(x)')
plt.plot(x_vals, V_vals[:, 1], label='V_z(x)')
plt.title('Skerkraftsferlar V(x) í y og z stefnur')
plt.xlabel('x (m)')
plt.ylabel('Skerkraftur (N)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.subplot(2, 1, 2)
plt.plot(x_vals, M_vals[:, 0], label='M_y(x)')
plt.plot(x_vals, M_vals[:, 1], label='M_z(x)')
plt.title('Beygjuvægi M(x) um y og z ás')
plt.xlabel('x (m)')
plt.ylabel('Beygjuvægi (Nm)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### (d) Hæstu beygjuspennur
Finnum heildar beygjuvægi í bæði $xy$ og $xz$ plönum með því að taka normið af vigurvægjunum úr Macauley fallinu $M(x)$ úr a lið. Hæsta beygjuvægið fæst svo með því að finna hámarkið á heildar vægjunum yfir bilið $x \in [0, x_C]$. Með Python getum við notað *numpy* pakkann til að reikna normið og *scipy.optimize* pakkann til að finna hámarkið. Hér eru niðurstöður fyrir hæstu beygjuspennur í bæði $xy$ og $xz$ plönum.

```python
#| echo: true
#| output: true

# skilgreinum normið af M(x) um y og z ás
def M_norm(x, R_1, R_2):
    M = M_fun(x, R_1, R_2)
    return np.sqrt(M[0]**2 + M[1]**2)

# finnum hámarkið á M_norm yfir bilið [0, x_C]
from scipy.optimize import minimize_scalar
res = minimize_scalar(
    lambda x: -M_norm(x, R_1, R_2), 
    bounds=(0, x_C), 
    method='bounded'
)
x_max = res.x
M_max = M_norm(x_max, R_1, R_2)
print(f"Hæsta M_norm er {M_max:.2f} Nm við x = {x_max:.2f} m")
print(f"Hæsta begyjuvægi um y ás er M_y = {M_fun(x_max, R_1, R_2)[0]:.2f} Nm")
print(f"Hæsta begyjuvægi um z ás er M_z = {M_fun(x_max, R_1, R_2)[1]:.2f} Nm")
print(
    f"Normið af hæstu beygjuvægi er {
        np.sqrt(
            M_fun(x_max, R_1, R_2)[0]**2 
            + M_fun(x_max, R_1, R_2)[1]**2
            ):.2f} Nm"
)
# teiknum M(x) um y og z ása ásamt norminu á 
# sama grafi með lóðrétta línu við x_max 
plt.figure(figsize=(10, 6))
plt.plot(x_vals, [M_norm(x, R_1, R_2) for x in x_vals], label='||M(x)||')
plt.plot(x_vals, [M_fun(x, R_1, R_2)[0] for x in x_vals], label='M_y(x)')
plt.plot(x_vals, [M_fun(x, R_1, R_2)[1] for x in x_vals], label='M_z(x)')
plt.axvline(x=x_max, color='red', linestyle='--', label='x_max')
plt.title('Normið af M(x) og M(x) um y og z ás')
plt.xlabel('x (m)')
plt.ylabel('Beygjuvægi (Nm)')
plt.legend()
plt.grid()
plt.show()
```

Þar sem um er að ræða tvívíða beygju um $y$ og $z$ ás, þá getum ákvarðara halla núllássins í gegnum þversniðið, þ.e. línan þar sem beygjuspennan er núll. Halli núllássins $\alpha$ er gefinn með
$$
    \tan{\alpha} = \frac{-M_yI_z}{M_zI_y},
$$þar sem þversniðið hefur einungis einn samhverfuás, þ.e. $y$ ásinn. Ákvörðum $I_y$ og $I_z$ fyrir gefið þversnið með eftirfarandi tölulegum útreikningum og ákvörðum einnig hornið $\alpha$.

```python
#| echo: true
#| output: true

# lóðréttur flangsi (#1)
b_1 = 22.225e-3 # m
h_1 = 63.5e-3 # m
y_1 = h_1/2 # m
A_1 = b_1 * h_1 # m^2
I1_y = (h_1*b_1**3) / 12 # m^
I1_z = (b_1 * h_1**3) / 12 # m^4

# láréttur flangsi (#2)
b_2 = 101.600e-3 # m
h_2 = 22.225e-3 # m
y_2 = h_2/2 + h_1 # m
A_2 = b_2 * h_2 # m^2
I2_y = (h_2*b_2**3) / 12 # m^4
I2_z = (b_2 * h_2**3) / 12 # m^4

# flatarmiðja (e. centroid)
y_c = (A_1*y_1 + A_2*y_2) / (A_1 + A_2)
print(f"Flatarmiðja y_c = {y_c*1e3:.2f} mm frá neðri brún")

# flatartregður Iy og Iz
I_y = I1_y + I2_y
I_z = (
    I1_z + A_1 * (y_c - y_1) ** 2 
    + I2_z + A_2 * (y_c - y_2) ** 2
)
print(f"Flatartregða Iy = {I_y*1e12:.3e} m^4")
print(f"Flatartregða Iz = {I_z*1e12:.3e} m^4")

# halli núllássins er
alpha = np.atan(-M_fun(x_max, R_1, R_2)[0]*I_z/(M_fun(x_max, R_1, R_2)[1]*I_y))
print(f"Halli núllássins alpha = {np.degrees(alpha):.2f}° frá +z ás") 
```

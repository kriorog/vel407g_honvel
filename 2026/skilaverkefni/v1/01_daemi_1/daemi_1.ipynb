# Dæmi 1
Gegnheill öxull með þvermál $d$ verður fyrir bæði jafndreifðri normalspennu $\sigma_n$ vegna áskrafts $P$ sem og beygjuspennu $\sigma_b$ vegna vægis $M$. Hæsta spenna á ytra yfirborði öxuls er $\sigma_{max}=\sigma_n+\sigma_b$.

Áskraftur $P$ hefur meðaltal $\mu_P=779 \text{ kN}$ og staðalfrávik $\hat{\sigma}_P = 72.7\text{ kN}$. Beygjuvægið hefur meðaltal $\mu_M=43.5 \text{ kNm}$ og staðalfrávik $\hat{\sigma}_M=2.53 \text{ kNm}$.

Öxull er úr stáli sem hefur meðalflotstyrk $\bar{S}_y = 553 \text{ MPa}$ og staðalfrávik $\hat{\sigma}_{S_y}=42.7 \text{ MPa}$. 

Gerið ráð fyrir að allar stærðir séu óháðar og normaldreifðar.

Ákvarðið hönnunarstuðul og þvermál öxuls m.v. 99.0\% áreiðanleika (öryggi) gegn flotbjögun.

## Lausn
Normalspennan er $\sigma_n = P/A$ og hæsta beygjuspenna er $\sigma_b = Mr/I = Md/2I$ þar sem $r=d/2$ er geisli öxuls. Þar sem hvor spennan fyrir sig er línulegt fall af normaldreifðri slembibreytu, þ.e. $P$ og $M$, þá eru spennurnar einnig normaldreifðar. 

Normalspennan $\sigma_n$ er því normaldreifð með meðaltal $\mu_{\sigma_n} = \mu_{P}/A$ og staðalfrávik $\hat{\sigma}_{\sigma_n} = \hat{\sigma}_{P}/A$. Hæsta beygjuspennan er einnig normaldreifð með meðaltal $\mu_{\sigma_b}=\mu_{M}d/2I$ og staðalfrávik $\hat{\sigma}_{\sigma_b}=\hat{\sigma}_M d/2I$.

Einnig er hæsta spenna $\sigma_{max}$ normaldreifð slembibreyta því hún er summa $\sigma_n$ og $\sigma_b$. Hæsta spennan hefur því meðaltal 
$$
\begin{aligned}
    \mu_{\sigma_{max}}  &= \mu_{\sigma_n} + \mu_{\sigma_b}, \\
                        &= \frac{\mu_P}{A} + \frac{\mu_M d}{2I},
\end{aligned}
$$
og staðalfrávik
$$
\begin{aligned}
    \hat{\sigma}_{\sigma_{max}} &= \sqrt{\left(\frac{\hat{\sigma}_P}{A}\right)^2 + \left(\frac{\hat{\sigma}_Md}{2I}\right)^2}.
\end{aligned}
$$

Fyrir hringlaga gegnheilt þversnið með þvermál $d$ er $A=\frac{\pi d^2}{4}$ og $I=\frac{\pi d^4}{64}$. Getum því ritað $I = \frac{d^2}{16}A$. Þar með má umrita meðaltalið og staðalfrávikið
$$
    \mu_{\sigma_{max}} = \frac{1}{A}\left( \mu_P + \frac{8 \mu_M}{d} \right),
$$
og
$$
    \hat{\sigma}_{\sigma_{max}} = \frac{1}{A}\sqrt{\hat{\sigma}_P^2 + \left(\frac{8\hat{\sigma}_M}{d}\right)^2}.
$$

:::{.callout-note title="Meðhöndlun á normaldreifðum slembibreytum"}
Til að staðfesta þessar niðurstöður[^1], þá þarf að athuga að líkindadreifingar $\sigma_n$ og $\sigma_b$ eru fengnar með því að skala normaldreifingar á óháðu slembibreytunum $P$ og $M$ með fasta. Ef slembibreyta $X$ er normaldreifð, þá er breytan $aX$ fyrir eitthvað $a \in \mathbb{R}$ einnig normaldreifð með meðaltal $\mu_{aX} = a\mu_{X}$ og staðalfrávik $\hat{\sigma}_{aX}=|a|X$. Ef $a \geq 0$ þá má sleppa tölugildinu $|\cdot|$ í staðalfrávikinu.

Ef $Y$ er önnur normaldreifð slembibreyta óháð $X$, þá er $Z=X+Y$ einnig normaldreifð slembibreyta með meðaltal $\mu_z = \mu_x + \mu_z$ og staðalfrávik $\hat{\sigma}_Z=\sqrt{\hat{\sigma}_{\sigma_n}^2+ \hat{\sigma}_{\sigma_b}^2}$.

[^1]: Sjá til dæmis Ross *Introduction to Probability and Statistics* kafla 4.5 og 4.6 úr STÆ203G "Lík og töl".
:::

Skoðum næst umframstyrkinn $m = S_y - \sigma_{max}$ (g.r.f að $\sigma_{max}$ sé stranglega jákvæð stærð), sem er summa tveggja normaldreifðra slembibreytna og er $m$ því normaldreift með meðaltal $\mu_m = \bar{S}_y - \mu_{\sigma_{max}}$ og staðalfrávik $\hat{\sigma}_m = \sqrt{\hat{\sigma}_{\sigma_{max}}^2 + \hat{\sigma}_{S_y}^2}$. Við krefjumst þess að áreiðanleikinn sé $R=p(m>0)=0.990$ sem má einnig rita sem $R=p(m>0)=1-P(m \leq 0)=1-p_f$ þar sem $p_f = P(m \leq 0)$ eru líkur á bilun/broti. Áreiðanleiki $R=0.990$ er því jafngilt $p_f=0.010$.

Látum $z_{0.010}$ vera z-gildið þannig að $P(Z \leq z_{0.010})=p_f=0.010$ fyrir staðlaða normaldreifingu $Z$. Hægt er að lesa gildið á $z_{0.010}$ út úr töflum eða reikna gildið $z_{0.010}=\Phi^{-1}(0.990)$ með $\texttt{scipy.stats.norm}$ pakkanum í python. Fallið $\texttt{norm.ppf}$ er fallið $\Phi^{-1}(p)$ sem andhverfa líkindadreifinafallsins $\Phi$ (ppf=percent point function) sem skilar z-gildinu fyrir gefnar líkur $p$. 

Líkurnar á broti $p_f$ eru
$$
\begin{aligned}
    p_f 
    &= P(m \leq 0) \\
    &= P\left(\frac{m - \mu_m}{\hat{\sigma}_m} \leq \frac{0 - \mu_m}{\hat{\sigma}_m}\right) \\
    &= P\left(Z \leq -\frac{\mu_m}{\hat{\sigma}_m}\right) \\
    &= P\left(Z \leq z_{0.010}\right).
\end{aligned}
$$
Við þurfum því að leysa jöfnuna
$$
    -\frac{\mu_m}{\hat{\sigma}_m} = z_{0.01}
$$
sem er jafngilt
$$
\begin{aligned}
    0 &= \mu_m + z_{0.01} \hat{\sigma}_m, \\
    &= \bar{S}_y - \mu_{\sigma_{max}} + z_{0.01} \hat{\sigma}_m, \\
    &= \bar{S}_y - \frac{1}{A}\left( \mu_P + \frac{8 \mu_M}{d} \right) 
        + z_{0.01} \sqrt{\frac{1}{A^2}\left(\hat{\sigma}_P^2 + \left(\frac{8\hat{\sigma}_M}{d}\right)^2\right) + \hat{\sigma}_{S_y}^2}.
\end{aligned}
$$

Við getum leyst þessa jöfnu tölulega fyrir þvermálið $d$ með því að finna núllstöðvar hennar (e. root finding).

```python
#| echo: true
#| output: true

from scipy.stats import norm
import numpy as np

# áreiðanleiki
r = 0.99 
z = norm.ppf(1-r)

# meðaltal og staðalfrávik fyrir álag og vægi
mu_P = 779e3 # N
stdev_P = 72.7e3 # N
mu_M = 43.5e3 # Nm
stdev_M = 2.53e3 # Nm

# meðaltal og staðalfrávik fyrir styrk
mu_strength = 553e6 # Pa
stdev_strength = 42.7e6 # Pa

# skilgreinum föll til að smíða jöfnuna sem á að leysa
A_fun = lambda d: np.pi * (d/2)**2  # þversniðsflatarmál, m2
mu_sigma_fun = lambda d: (mu_P + 8*mu_M/d) / A_fun(d) # meðaltal sigma_max
mu_m_fun = lambda d: mu_strength - mu_sigma_fun(d) # metaltal umframstyrks m
stdev_m_fun = lambda d: np.sqrt(
        (stdev_P**2 + (8*stdev_M/d)**2) / A_fun(d)**2
        + stdev_strength**2
) # staðalfrávik umframstyrks m

# fallið með núllstöðvar sem við viljum finna.
d_fun = lambda d: mu_m_fun(d) + z * stdev_m_fun(d) 

# teiknum d_fun til að sjá hegðun fallsins (og það hafi núllstöðvar)
import matplotlib.pyplot as plt
d_values = np.linspace(50, 200, 400)/1000 # í metrum
d_fun_values = [d_fun(d) for d in d_values]
plt.plot(d_values, d_fun_values)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Þvermál d (mm)')
plt.ylabel('d_fun(d)')
plt.title('Hegðun d_fun sem fall af þvermáli d')
plt.grid()
plt.show()

# leysum jöfnuna með því að finna núllstöðvar
print(f"z-gildi fyrir áreiðanleika {r*100}% eða bilanalíkur {100-r*100}% er: {z:.4f}")

from scipy.optimize import fsolve
d_0 = 10/1000 # m, upphafsgisk
d_solution = fsolve(d_fun, d_0)
d_final = d_solution[0]
print(f"Þvermál d til að ná áreiðanleika {r*100}% er: {d_final*1000:.2f} mm")

z_final = -mu_m_fun(d_final) / stdev_m_fun(d_final)
R_final = 1 - norm.cdf(-mu_m_fun(d_final) / stdev_m_fun(d_final))
print(f"Staðfesting: Áreiðanleiki R = {R_final:.3f} (ætti að vera {r:.3f})")
```

Þvermálið er því $d=108 \text{ mm}$ til að ná áreiðanleika $R=0.990$.

Fyrir gefið $d=d_f$ sem uppfyllir áreiðanleikakröfu, þá er hönnunarstuðullinn gefinn sem hlutfall hönnunarstyrks og meðaltals á hæstu spennu
$$
    n   = \frac{\bar{S}_y}{\mu_{\sigma_{max}}} 
        = \frac{\bar{S}_y}{\frac{1}{A}\left( \mu_P + \frac{8 \mu_M}{d_f} \right)}
        = \frac{\bar{S}_y A}{\mu_P + \frac{8 \mu_M}{d_f}}.
$$

```python
#| echo: true
#| output: true

# reiknum hönnunarstuðul
nd = mu_strength / mu_sigma_fun(d_final)
print(f"Hönnunarstuðull nd = {nd:.2f}")
```

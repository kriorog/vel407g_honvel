# Dæmi 1
Gegnheill öxull með þvermál $d$ verður fyrir bæði jafndreifðri normalspennu $\sigma_n$ vegna áskrafts $P$ sem og beygjuspennu $\sigma_b$ vegna vægis $M$. Hæsta spenna á ytra yfirborði öxuls er $\sigma_{max}=\sigma_n+\sigma_b$.

Áskraftur $P$ hefur meðaltal $\mu_P=86.6 \text{ kN}$ og staðalfrávik $\hat{\sigma}_P = 8.08 \text{ kN}$. Beygjuvægið hefur meðaltal $\mu_M=1610 \text{ kNm}$ og staðalfrávik $\hat{\sigma}_M=93.9 \text{ MNm}$.

Öxull er úr stáli sem hefur meðalflotstyrk $\bar{S}_y = 553 \text{ MPa}$ og staðalfrávik $\hat{\sigma}_{S_y}=42.7 \text{ MPa}$. 

Gerið ráð fyrir að allar stærðir séu óháðar og normaldreifðar.

Ákvarðið hönnunarstuðul og þvermál öxuls m.v. (a) 95% og (b) 99% áreiðanleika gegn floti.

```python
import math

# converting the given stresses to force and moment
d = 35 # mm
A = math.pi*d**2/4
I = math.pi*d**4/64

# mean stresses
mu_a = 90 # MPa
mu_b = 383 # MPa

# stdev
stdev_a = 8.4 # MPa
stdev_b = 22.3 # MPa

# convert axial stress to load P [kN]
mu_p = mu_a*A/1000
stdev_p = stdev_a*A/1000

# convert bending stress to moment M [kNm]
mu_m = mu_b*I/(d/2)/1000
stdev_m = stdev_b*I/(d/2)/1000

print(f"Cross section area is {A:.3f} mm^2")
print(f"Moment of inertia is {I:.3f} mm^4")
print(f"Mean force P is {mu_p: .2f} kN")
print(f"Force P stdev is {stdev_p: .3f} kN")
print(f"Mean moment M is {mu_m: .3f} kNm")
print(f"Moment stdev is {stdev_m: .3f} kNm")
```
